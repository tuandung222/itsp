# Cách sử dụng sách này

Điểm khởi đầu được khuyến nghị là trang web [https://speechprocessingbook.aalto.fi/](https://speechprocessingbook.aalto.fi/). Trang web này giúp truy cập hầu hết nội dung cuốn sách một cách rất dễ dàng. 

Để chạy thử các ví dụ mã nguồn và trải nghiệm đầy đủ các minh họa tương tác, chúng tôi cung cấp các tùy chọn sau:

1. Trực tiếp từ [trang web](https://speechprocessingbook.aalto.fi/) với các ví dụ mã nguồn và thành phần tương tác thông qua [Mybinder](https://mybinder.org/) hoặc [Colab](https://colab.research.google.com/). (Lưu ý: Mybinder thường khởi động chậm và trên Colab một số notebook hiện tại có thể không hoạt động.) Để sử dụng tính năng này, hãy nhấp vào logo hình tên lửa ở đầu trang. 
2. Tải sách từ GitHub [https://github.com/Speech-Interaction-Technology-Aalto-U/itsp](https://github.com/Speech-Interaction-Technology-Aalto-U/itsp). Chạy JupyterLab trên một [máy chủ nội bộ (local server)](content:localserver).

Lợi ích của việc tự chạy mã nguồn bao gồm:
- Bạn có thể thay đổi các tham số và mã nguồn để quan sát sự thay đổi hành vi tương ứng.
- Bạn có thể sử dụng các tệp âm thanh của riêng mình (chẳng hạn như ghi âm giọng nói của chính bạn) trong cả các ví dụ mã nguồn và trực quan hóa tương tác.

## Sử dụng máy chủ nội bộ (local server)

Mặc dù không bắt buộc đối với hầu hết độc giả, để có quyền truy cập toàn diện vào tất cả nội dung, bạn có thể chạy một máy chủ [Jupyter](https://jupyter.org/) nội bộ để truy cập cuốn sách này. Thực hiện theo các bước dưới đây để thiết lập.

1. Tải sách từ GitHub [https://github.com/Speech-Interaction-Technology-Aalto-U/itsp](https://github.com/Speech-Interaction-Technology-Aalto-U/itsp).

2. Cài đặt [Anaconda](https://anaconda.org) (khuyến nghị) hoặc dịch vụ quản lý gói khác để cài đặt [Python](https://www.python.org/). 

3. Khuyến nghị (nhưng có thể không bắt buộc) cấu hình Anaconda bằng cách thêm các kênh (channels) sau.

   ```bash
   conda config --add channels conda-forge
   conda config --add channels pytorch
   ```
   ![]()
   Sau đó các kênh này sẽ xuất hiện trong tệp `.condarc` của bạn cùng với `default`.

4. Cài đặt các gói phần mềm sau
   ```bash
   conda install matplotlib scipy numpy jupyter jupyterlab torch torchaudio ipython ipywidgets
   ```
    ![]()
5. Trong thư mục lưu trữ sách, khởi chạy JupyterLab bằng lệnh
   ```bash
   jupyter lab &
   ```   
   ![]()
6. Mở tệp `index.md` và bắt đầu đọc.

## Xuất bản dưới dạng PDF hoặc EPUB

Để đọc tài liệu này trên thiết bị di động hoặc in toàn bộ ra giấy, việc xuất bản cuốn sách sang định dạng PDF hoặc EPUB có thể rất hữu ích. Chúng tôi không hỗ trợ chính thức tính năng xuất bản này (nghĩa là nó có thể không hoạt động hoàn hảo), nhưng đây là phương pháp đã được báo cáo là hoạt động thành công đối với một số người dùng.

1. Tải sách từ GitHub [https://github.com/Speech-Interaction-Technology-Aalto-U/itsp](https://github.com/Speech-Interaction-Technology-Aalto-U/itsp).

2. Cài đặt [Anaconda](https://anaconda.org) (khuyến nghị) hoặc dịch vụ quản lý gói khác để cài đặt [Python](https://www.python.org/). 

3. Khuyến nghị (nhưng có thể không bắt buộc) cấu hình Anaconda bằng cách thêm các kênh sau.

   ```bash
   conda config --add channels conda-forge
   conda config --add channels pytorch
   ```
   ![]()   
   Sau đó các kênh này sẽ xuất hiện trong tệp `.condarc` của bạn cùng với `default`.

4. Cài đặt các gói phần mềm sau

   ```bash
   conda install matplotlib scipy numpy jupyter jupyterlab torch torchaudio ipython ipywidgets jupyterbook
   ```
    ![]()    
    Lưu ý rằng khác với phần [Máy chủ nội bộ](content:localserver), ở đây chúng ta cần cài đặt thêm gói `jupyterbook`.
    
5. Di chuyển đến thư mục chứa kho lưu trữ git đã tải về và biên dịch toàn bộ cuốn sách thành một tệp HTML duy nhất bằng lệnh 

   ```bash 
   jb build . --builder singlehtml
   ```
   ![]()
    
5. Cài đặt [Pandoc](https://pandoc.org/) và sử dụng nó để chuyển đổi tệp HTML sang định dạng PDF hoặc EPUB.

Các hướng dẫn này được tổng hợp từ cuộc thảo luận trên [GitHub](https://github.com/Speech-Interaction-Technology-Aalto-U/itsp/discussions/1#discussioncomment-5553785), nơi cũng chứa một số phương pháp thay thế khác.

# Sử dụng sách này làm tài liệu giảng dạy

Tài liệu này được thiết kế cho mục đích giảng dạy, tự học cũng như phục vụ các khóa học. Để sử dụng làm tài liệu cho khóa học, chúng tôi nhận thấy việc tạo một trang mục lục mới cho mỗi khóa học là rất hữu ích, qua đó lựa chọn và liên kết đến các chương và phần có liên quan. Chúng tôi cũng sử dụng hệ thống Jupyter hub riêng của khóa học trên mạng nội bộ (Intranet), nhờ đó sinh viên có thể chạy trực tiếp tài liệu tại đó mà không cần cài đặt bất kỳ môi trường nào và tránh được rủi ro các tài nguyên miễn phí như MyBinder phản hồi chậm.
