# Sản sinh tiếng nói và các tính chất âm học


## Sản sinh tiếng nói về mặt sinh lý học

### Tổng quan

Khi một người có ý muốn hoặc dự định nói, não bộ sẽ hình thành một câu
với ý nghĩa dự định và ánh xạ chuỗi từ thành các chuyển động sinh lý cần
thiết để tạo ra chuỗi âm thanh tiếng nói tương ứng. Phần thần kinh của quá
trình sản sinh tiếng nói không được thảo luận thêm ở đây.

Hoạt động thể chất bắt đầu bằng việc co cơ phổi, đẩy không khí từ phổi ra
ngoài, đi qua cổ họng, khoang miệng và khoang mũi. Bản thân luồng khí không
tạo ra âm thanh nghe được — âm thanh là sự dao động của áp suất không khí.
Do đó, để tạo ra âm thanh, chúng ta cần cản trở luồng khí để tạo ra dao
động hoặc nhiễu loạn. Các dao động chủ yếu được tạo ra khi [*dây thanh
*](https://en.wikipedia.org/wiki/Vocal_cords) (vocal folds) được căng ở mức
phù hợp. Điều này tạo ra *âm hữu thanh* (voiced sounds) và có lẽ là tính
chất đặc trưng nhất của tín hiệu tiếng nói. Dao động cũng có thể được tạo ra
bởi các bộ phận khác của cơ quan sản sinh tiếng nói, chẳng hạn như lưỡi dao
động trên răng tạo ra âm /r/ rung, hoặc tiểu thiệt dao động trong luồng khí,
được gọi là rung tiểu thiệt (cụ thể là một dạng /r/ hầu). Tuy nhiên, những
rung động này, cả bằng lưỡi và tiểu thiệt, không nên nhầm lẫn với âm hữu
thanh, vốn luôn được tạo ra bởi dao động của dây thanh. Các âm không có dao
động ở dây thanh được gọi là *âm vô thanh* (unvoiced sounds).

Hầu hết các âm vô thanh điển hình được gây ra bởi nhiễu loạn do sự thu hẹp
tĩnh của luồng khí tại bất kỳ phần nào của các khoang không khí phía trên
dây thanh (cụ thể là thanh quản, hầu và khoang miệng hoặc khoang mũi). Ví
dụ, khi đặt lưỡi gần răng, ta thu được phụ âm /s/, và bằng cách chặn rồi
thả luồng khí thông qua đóng mở môi, ta thu được phụ âm /p/. Một lớp âm vị
đặc biệt nữa là các phụ âm mũi, trong đó luồng khí qua miệng bị chặn hoàn
toàn hoặc một phần, sao cho phần lớn không khí đi qua mũi.

### Dây thanh

*Dây thanh* (vocal folds), còn gọi là thanh đới, nằm trong cổ họng và dao
động để tạo ra âm hữu thanh. Khe hở giữa hai dây thanh (khoảng không gian
trống giữa các dây thanh) được gọi là *thanh môn* (glottis). Tương ứng,
khoảng không gian giữa dây thanh và phổi được gọi là vùng *dưới thanh môn*
(subglottal).

Khi áp suất dưới thanh môn, hay *áp suất dưới thanh môn* (subglottal
pressure), tăng lên, nó đẩy dây thanh mở ra. Khi mở, không khí tràn qua dây
thanh. Chuyển động ngược lại, đóng dây thanh, chủ yếu được gây ra bởi [hiệu
ứng
Venturi](https://en.wikipedia.org/wiki/Venturi_effect), tạo ra sự giảm áp
suất không khí giữa các dây thanh khi không khí chảy qua chúng. Khi dây
thanh đóng lại, cuối cùng chúng sẽ va chạm vào nhau. Sự dừng đột ngột của
luồng khí này là sự kiện âm học lớn nhất tại dây thanh và được gọi là *kích
thích thanh môn* (glottal excitation).

Về mặt luồng khí, kết quả là trong *pha đóng* (closed phase) (khi dây thanh
đóng), không có luồng khí nào. Khi bắt đầu *pha mở* (open phase) (khi dây
thanh mở), không khí bắt đầu chảy qua thanh môn và rõ ràng khi dây thanh
đóng lại thì luồng khí cũng giảm. Tuy nhiên, do quán tính của bản thân
không khí, chuyển động của không khí xảy ra hơi trễ so với chuyển động của
dây thanh. Nói cách khác, có một sự chênh lệch pha giữa chuyển động dây
thanh và dạng sóng luồng khí thanh môn.

Tần số dao động của dây thanh phụ thuộc vào ba thành phần chính: mức độ
căng dọc của dây thanh, chênh lệch áp suất trên và dưới dây thanh, cũng
như chiều dài và khối lượng của dây thanh. Áp suất và độ căng có thể được
chủ động thay đổi để thay đổi tần số. Chiều dài và khối lượng dây thanh
tương quan với kích thước cơ thể của người nói, điều này giải thích tại sao
trẻ em và phụ nữ trung bình có cao độ (pitch) cao hơn nam giới.

Lưu ý rằng tần số của dây thanh chỉ hiện tượng vật lý thực tế, trong khi
*cao độ* (pitch) chỉ sự cảm nhận tần số. Có nhiều trường hợp hai khái niệm
này khác nhau, ví dụ, cộng hưởng trong đường dẫn âm (vocal tract) có thể
nhấn mạnh các họa âm của tần số cơ bản (fundamental frequency) sao cho các
họa âm vang hơn tần số cơ bản, và ta cảm nhận một trong các họa âm đó là
tần số cơ bản. Cao độ cảm nhận khi đó là tần số của họa âm thay vì tần số
cơ bản.

### Đường dẫn âm

Đường dẫn âm (vocal tract), bao gồm thanh quản, hầu và các khoang miệng, có
ảnh hưởng lớn đến âm sắc của âm thanh. Cụ thể, hình dạng của đường dẫn âm
quyết định các cộng hưởng và phản cộng hưởng của không gian âm học, làm
tăng cường và suy giảm các tần số khác nhau của âm thanh. Hình dạng được
quyết định bởi nhiều thành phần, đặc biệt là vị trí của hàm, môi và lưỡi.
Các cộng hưởng dễ dàng được người nói thay đổi và người nghe cảm nhận, do
đó có thể được sử dụng trong giao tiếp để truyền tải thông tin. Cụ thể, các
đặc trưng âm học phân biệt các *nguyên âm* (vowels) với nhau là các tần số
cộng hưởng trong đường dẫn âm, tương ứng với các *vị trí* cấu âm cụ thể chủ
yếu về mặt vị trí lưỡi. Do không khí có thể lưu thông tương đối không bị
cản trở, các nguyên âm thường có năng lượng và độ lớn cao hơn so với *phụ
âm* (consonants).

Trong các âm *phụ âm*, có sự cản trở một phần hoặc toàn bộ tại một vị trí
nào đó trong đường dẫn âm. Ví dụ, *phụ âm xát* (fricative consonants) đặc
trưng bởi một khe hẹp giữa lưỡi và phía trước/trên của miệng, dẫn đến luồng
khí hỗn loạn giống tiếng rít. Trong âm tắc (plosives), luồng khí trong
đường dẫn âm bị chặn hoàn toàn tạm thời. Ví dụ, *âm tắc hai môi* (bilabial
plosives) đặc trưng bởi việc đóng kín môi tạm thời, dẫn đến tích tụ áp suất
không khí trong đường dẫn âm do áp suất phổi duy trì. Khi môi mở ra, không
khí tích tụ được giải phóng cùng với một âm *bùng nổ* ngắn (plosion) có đặc
tính xung và nhiễu. Tương tự như nguyên âm, *vị trí* cản trở trong miệng
(tức vị trí cấu âm) sẽ ảnh hưởng đến đặc tính âm học của phụ âm bằng cách
thay đổi đặc tính âm học của đường dẫn âm. Ngoài ra, *phương thức* cấu âm
được dùng để đặc trưng các phụ âm khác nhau, vì có nhiều cách tạo ra tiếng
nói trong khi vị trí cản trở chính có thể giữ nguyên (ví dụ: các *âm vỗ*
ngắn và *âm lật*, các *rung* lặp lại, hoặc các khe hẹp đã đề cập cho *âm
xát*).

Về mặt hình dạng đường dẫn âm, một lớp phụ âm đặc biệt là *âm mũi*
(nasals), được tạo ra khi màn hầu (cấu trúc mềm ở phía sau trên của khoang
miệng) mở, cho phép không khí đi vào *khoang mũi*. Khi màn hầu mở, đường
dẫn âm có thể được xem như một ống chung từ thanh quản đến phía sau miệng,
sau đó đường dẫn chia thành hai nhánh song song gồm khoang miệng và khoang
mũi. Việc ghép nối khoang mũi vào đường dẫn âm có ảnh hưởng rõ rệt đến các
cộng hưởng và phản cộng hưởng của đường dẫn. Điều này thường được người
nghe cảm nhận như sự *mũi hóa* (nasalization) của âm thanh tiếng nói.

## Các tính chất âm học của tín hiệu tiếng nói

Các đặc trưng âm học quan trọng nhất của tín hiệu tiếng nói là (nói một
cách khái quát)

-   *Cộng hưởng của đường dẫn âm*, đặc biệt là hai cộng hưởng thấp nhất,
    được gọi là các *formant* (cộng hưởng đường dẫn âm) F1 và F2 (xem hình
    bên dưới). Cấu trúc cộng hưởng có thể dễ dàng khảo sát bằng cách vẽ một
    "*bao*" phía trên phổ, tức là vẽ một đường mượt đi ngay phía trên phổ,
    như hình bên dưới. Ta thu được *bao phổ* (spectral envelope), đặc trưng
    cho hình dạng vĩ mô của phổ tín hiệu tiếng nói và thường được dùng để
    mô hình hóa tín hiệu tiếng nói.
-   *Tần số cơ bản* (fundamental frequency) của tín hiệu tiếng nói hoặc sự
    vắng mặt của nó mang rất nhiều thông tin. Theo định nghĩa, âm vị hữu
    thanh và vô thanh lần lượt là những âm có hoặc không có dao động ở dây
    thanh. Do tính nổi bật của nó, ta phân loại âm vị theo tiêu chí hữu
    thanh hay vô thanh.  
    Luồng khí đi qua dây thanh rung tạo ra dạng sóng giống như [hình sin
    chỉnh lưu nửa
    sóng](https://en.wikipedia.org/wiki/Rectifier#Half-wave_rectification).
    Tức là, luồng khí bằng không khi dây thanh đóng (pha đóng) và trong
    thời gian mở (pha mở) dạng sóng gần giống với hình dạng phần trên của
    một sóng sin. Phổ của dạng sóng này sẽ có cấu trúc của một tín hiệu
    điều hòa, tức là phổ sẽ có các đỉnh tại tần số cơ bản và các bội số
    nguyên của nó (xem hình bên dưới).  
    Ở hầu hết các ngôn ngữ, cao độ không phân biệt giữa các âm vị. Tuy
    nhiên, trong các ngôn ngữ được gọi là ngôn ngữ *thanh điệu*, đường bao
    cao độ theo thời gian mang ý nghĩa ngữ nghĩa (xem
    [Wikipedia:Tone
    (linguistics)](https://en.wikipedia.org/wiki/Tone_(linguistics)) để có
    mẫu âm thanh minh họa). Tuy nhiên, đường bao cao độ thường được dùng để
    mã hóa sự *nhấn mạnh* trong câu. Nói một cách khái quát, việc tăng nỗ
    lực vật lý lên một âm vị sẽ nâng cao độ và cường độ của nó, và điều đó
    thường được hiểu là nhấn mạnh, tức là từ (hoặc âm vị) được nhấn mạnh
    quan trọng hơn các từ (hoặc âm vị) khác trong câu.
-   *Biên độ* hoặc *cường độ* tín hiệu theo thời gian là một đặc trưng quan
    trọng khác và ở dạng thô nhất có thể là sự khác biệt giữa tiếng nói và
    im lặng (xem thêm [Phát hiện hoạt động giọng nói
    (VAD)](content:vad)). Hơn nữa, có những âm vị được đặc trưng bởi cấu
    trúc thời gian; đặc biệt là *âm tắc* và *âm nổ* (plosive-consonants),
    trong đó luồng khí bị chặn và sau đó được giải phóng (ví dụ: /p/, /t/
    và /k/). Mặc dù phần tắc không nghe rõ, chính sự tương phản giữa im
    lặng trước khi bùng nổ năng lượng mới là đặc điểm nổi bật của những phụ
    âm này.


Phổ của một đoạn tiếng nói được chú thích các formant $F_k$ (với $ k\geq 1 $) cùng với tần số cơ bản $F_0$ và
các bội số nguyên $kF_0$. Lưu ý rằng không phải lúc nào cũng xác định rõ
vị trí các formant; ở đây formant F4 và F5 không nổi bật và do đó khó xác
định.

## Mô hình hóa sinh lý học

### Đường dẫn âm

Nguyên âm có vai trò trung tâm trong giao tiếp bằng lời nói, và nguyên âm
được xác định bởi hình dạng của đường dẫn âm. Do đó, việc mô hình hóa đường
dẫn âm có ý nghĩa đặc biệt quan trọng.

#### Các mô hình đơn giản

Đường dẫn âm về cơ bản là một ống có chiều dài thay đổi. Nó có một khúc cong
90 độ, nơi cổ họng chuyển thành khoang miệng, nhưng ảnh hưởng âm học của
khúc cong này không đáng kể và có thể bỏ qua trong các mô hình đơn giản.
Ống này có hai đường dẫn, qua khoang miệng và khoang mũi. Ảnh hưởng âm học
của khoang miệng chi phối tín hiệu đầu ra sao cho, nói một cách khái quát,
khoang miệng tạo ra các cộng hưởng cho âm thanh đầu ra, trong khi khoang
mũi chủ yếu đóng góp các phản cộng hưởng (các điểm trũng hoặc thung lũng)
cho bao phổ. Sự hiện diện của năng lượng quan trọng hơn về mặt tri giác so
với sự vắng mặt của nó, do đó các phản cộng hưởng có thể được bỏ qua trong
các mô hình đơn giản.

Một mô hình rất đơn giản do đó là một ống trụ thẳng được chia thành các
đoạn có bán kính không đổi và chiều dài bằng nhau (xem hình minh họa bên
dưới). Nếu ta giả sử thêm rằng các đoạn ống này không tổn hao, thì ống này
tương đương giải tích với một [bộ dự báo tuyến
tính](content:linearprediction) (linear predictor). Đây là một sự đơn giản
hóa tuyệt vời ở chỗ từ một mô hình có cơ sở sinh lý học, ta thu được một
mô hình giải tích hợp lý mà các tham số có thể dễ dàng ước lượng từ tín
hiệu quan sát được. Thực tế, tương quan thời gian của tín hiệu tiếng nói
có thể được mô hình hóa rất hiệu quả bằng bộ dự báo tuyến tính. Nó mang
lại mối liên hệ hấp dẫn giữa mô hình sinh lý và mô hình tín hiệu. Đáng
tiếc, nó không hoàn toàn chính xác.

Mặc dù tín hiệu tiếng nói được mô hình hóa rất hiệu quả bởi bộ dự báo tuyến
tính, và bộ dự báo tuyến tính tương đương giải tích với mô hình ống, *các bộ
dự báo tuyến tính ước lượng từ tín hiệu âm thanh không nhất thiết tương ứng
với ống đã tạo ra âm thanh đó*. Sự không khớp về hình dạng giữa ống ước
lượng và ống thực tế xuất phát từ hai nguyên nhân chính;

1.  Việc ước lượng các hệ số dự báo tuyến tính (linear predictive
    coefficients) giả định rằng kích thích, tức kích thích thanh môn, là
    không tương quan (nhiễu trắng). Đây chắc chắn là một giả định sai. Mặc
    dù cấu trúc tuần hoàn của kích thích thanh môn không gây sai lệch đáng
    kể cho bộ dự báo tuyến tính, nhưng kích thích thanh môn cũng bị chi
    phối bởi các thành phần tần số thấp, sẽ làm lệch bộ dự báo tuyến tính.
    Bộ dự báo tuyến tính không thể phân biệt giữa các đặc điểm của kích
    thích thanh môn và đóng góp của đường dẫn âm, mà mô hình hóa cả hai một
    cách không phân biệt. Chúng ta cũng không biết đóng góp chính xác của
    kích thích thanh môn nên rất khó bù đắp cho nó.
2.  Mối quan hệ giải tích giữa các hệ số của bộ dự báo tuyến tính và bán
    kính của các đoạn mô hình ống có tính phi tuyến cao và nhạy cảm với sai
    số ước lượng. Sai số nhỏ trong tham số dự báo có thể dẫn đến những thay
    đổi lớn về hình dạng của mô hình ống.

Tuy nhiên, vì bộ dự báo tuyến tính hiệu quả trong việc mô hình hóa tiếng
nói, chúng vẫn hữu ích trong mô hình hóa tiếng nói ngay cả khi mối liên hệ
với mô hình ống dễ bị sai số. Dự báo tuyến tính (linear prediction) đặc
biệt hấp dẫn vì nó mang lại các thuật toán hiệu quả về mặt tính toán.

#### Các mô hình nâng cao

Khi cần mô hình hóa đường dẫn âm chính xác hơn, ta phải đánh giá lại các
giả định. Với [ống dẫn sóng số
](https://en.wikipedia.org/wiki/Digital_waveguide_synthesis) (digital
waveguides), ta có thể dễ dàng xây dựng các mô hình bao gồm đường dẫn thứ
hai tương ứng với đường mũi. Điểm xuất phát cho các mô hình như vậy là dự
báo tuyến tính, được biểu diễn dưới dạng đường trễ (delay-line) với các
phản xạ tương ứng tại các giao diện giữa các đoạn ống. Đường mũi sau đó có
thể được đưa vào bằng cách thêm một đường trễ thứ hai. Những mô hình này
hiệu quả về mặt tính toán trong việc tổng hợp âm thanh, nhưng việc ước
lượng tham số của chúng từ âm thanh thực tế có thể gặp khó khăn.

Nâng cao độ chính xác hơn nữa, ta tiến vào mô hình hóa vật lý đầy đủ như
[phương pháp phần tử hữu
hạn](https://en.wikipedia.org/wiki/Finite_element_method) (FEM). Tại đây,
ví dụ, thể tích không khí của đường dẫn âm có thể được chia thành các phần
tử nhỏ tương tác với nhau tuân theo [động lực học chất
lưu](https://en.wikipedia.org/wiki/Fluid_dynamics). Lưới phần tử càng dày
đặc thì mô hình càng tương ứng chính xác với thực tế vật lý. Việc đo lường
và mô hình hóa đường dẫn âm bằng phương pháp này rất phức tạp và là [một
lĩnh vực chuyên môn riêng](http://speech.math.aalto.fi/about.html).

### Hoạt động thanh môn

Để đặc trưng hóa lưu lượng thanh môn, ta định nghĩa các sự kiện trong một
chu kỳ thanh môn đơn lẻ như sau (minh họa trong hình bên dưới):

-   *Thời điểm mở và đóng* (hoặc thời điểm), là các điểm thời gian mà dây
    thanh mở và đóng tương ứng, và tại đó lưu lượng thanh môn bắt đầu và
    kết thúc.
-   *Pha mở và pha đóng*, là các khoảng thời gian mà thanh môn mở và đóng
    tương ứng.
-   Khoảng thời gian thanh môn mở và đóng được gọi tương ứng là *thời gian
    mở (OT)* và *thời gian đóng (CT)*. Do đó, độ dài chu kỳ là
    $T=OT+CT$.
-   *Pha mở rộng và pha thu hẹp* là các phần của pha mở, khi thanh môn đang
    mở rộng và thu hẹp tương ứng.
-   Độ dốc của pha thu hẹp liên quan đến "mức độ mạnh mẽ" của xung, tức là
    nó liên quan đến sức căng của dây thanh và được đặc trưng bởi *đỉnh âm
    của đạo hàm lưu lượng thanh môn* (giá trị âm).
-   Tất cả các tham số mô tả một khoảng thời gian thường được chuẩn hóa
    theo độ dài chu kỳ $t$.

Tương tự như mô hình hóa đường dẫn âm, trong mô hình hóa hoạt động thanh
môn cũng có một loạt các mô hình với độ phức tạp khác nhau:

-   *Dự báo tuyến tính pha cực đại* (Maximum-phase linear prediction); Sự
    kiện quan trọng nhất trong một xung lưu lượng thanh môn đơn lẻ là thời
    điểm đóng; dạng sóng trước đó mượt mà nhưng sự kiện đóng lại đột ngột.
    Dạng sóng do đó có thể được diễn giải như đáp ứng xung của một bộ lọc
    IIR nhưng đảo ngược, còn được gọi là đáp ứng xung của bộ dự báo tuyến
    tính pha cực đại (hình bên phải được tạo bằng phương pháp này). Ưu điểm
    của phương pháp này là nó tương tự với mô hình hóa đường dẫn âm bằng dự
    báo tuyến tính, nên ta đã quen thuộc với phương pháp và độ phức tạp tính
    toán thấp. Tuy nhiên, lưu ý rằng bộ lọc pha cực đại theo định nghĩa là
    không ổn định (không thể thực hiện được), nhưng ta luôn phải xử lý tín
    hiệu theo chiều ngược lại, điều này gây phức tạp cho thiết kế hệ thống.
-   *Mô hình Liljencrants-Fant (LF)* là một mô hình cổ điển về lưu lượng
    thanh môn, dạng ban đầu là một hàm bốn tham số ([được định nghĩa trong
    bài báo](http://www.speech.kth.se/prod/publications/files/qpsr/1985/1985_26_4_001-013.pdf)).
    Nó rất hữu ích và có ảnh hưởng vì nó tham số hóa lưu lượng bằng một số
    lượng nhỏ các tham số dễ hiểu. Đánh đổi là các tham số không dễ ước
    lượng từ tín hiệu thực tế và nó dựa trên bằng chứng giai thoại về hình
    dạng lưu lượng thanh môn, và nếu được đề xuất ngày nay, để được chấp
    nhận rộng rãi, chúng ta sẽ yêu cầu thêm bằng chứng hỗ trợ.
-   *[Hệ thống khối lượng-lò
    xo](https://en.wikipedia.org/wiki/Harmonic_oscillator#Spring/mass_system)*;
    các dây thanh đối diện có thể được mô hình hóa như các điểm khối lượng
    đơn giản được kết nối bằng lò xo giảm chấn tới các điểm cố định. Khi
    chịu lực Venturi tạo ra bởi luồng khí, các khối lượng này có thể dao
    động giống như dây thanh. Những mô hình này hấp dẫn vì, một lần nữa,
    tham số của chúng có ý nghĩa vật lý, nhưng vì tham số khó ước lượng từ
    dữ liệu thực tế và chúng chỉ dao động trong một phạm vi tham số hạn
    chế, khả năng ứng dụng thực tế bị giới hạn.
-   *[Phương pháp phần tử hữu hạn
    (FEM)](https://en.wikipedia.org/wiki/Finite_element_method)* một lần
    nữa là phương pháp cuối cùng cho phân tích chính xác, phù hợp ví dụ
    trong phân tích y khoa, tuy nhiên độ phức tạp tính toán quá lớn đối với
    các ứng dụng tiêu dùng.

### Bức xạ môi

Sau khi đi qua đường dẫn âm, không khí thoát ra chủ yếu qua miệng và một
phần qua mũi. Khi rời khỏi ống này, nó đi vào trường tự do nơi luồng khí
hầu như không có ảnh hưởng. Hãy nhớ lại rằng âm thanh thực chất là sự biến
thiên của áp suất không khí. Tại điểm chuyển tiếp từ ống sang trường tự do,
sự biến thiên của lưu lượng không khí trở thành sự biến thiên của áp suất
không khí.

Vật lý của hiện tượng này được chi phối bởi [động lực học chất
lưu](https://en.wikipedia.org/wiki/Fluid_dynamics), một chủ đề chuyên sâu,
nhưng về mặt trực quan ta có thể hình dung rằng sự biến thiên áp suất không
khí liên quan đến sự biến thiên của luồng khí. Do đó, nếu ta lấy đạo hàm
của luồng khí, ta được một xấp xỉ về ảnh hưởng của nó lên áp suất không khí

$$ 
sound(t) \approx \frac d{dt} flow(t), $$

trong đó $t$ là thời gian.

Thường ta làm việc với các tín hiệu được lấy mẫu tại các chỉ số thời gian
$n$, trong đó đạo hàm có thể được xấp xỉ thêm bằng sai phân bậc nhất

$$ 
sound(n) \approx g \left[flow(n) - flow(n-1)\right], 
$$

trong đó $g>0$ là một hệ số khuếch đại vô hướng.
