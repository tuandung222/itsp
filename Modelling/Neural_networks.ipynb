# Mạng nơ-ron

## Giới thiệu

Mạng nơ-ron nhân tạo (Artificial Neural Network - ANN) là một mô hình toán học mô phỏng cấu trúc và chức năng của mạng nơ-ron sinh học. Khối xây dựng cơ bản của mọi mạng nơ-ron nhân tạo là nơ-ron nhân tạo, tức là một mô hình toán học đơn giản
(hàm số). Nơ-ron nhân tạo là khối xây dựng cơ bản của mọi
mạng nơ-ron nhân tạo. Thiết kế và chức năng của nó được rút ra từ việc quan sát nơ-ron sinh học — khối xây dựng cơ bản của các mạng nơ-ron sinh học (hệ thống) bao gồm não bộ, tủy sống
và hạch ngoại vi. 


Mô hình tính toán sử dụng mạng nơ-ron được khởi nguồn từ những năm 1940
và đã trải qua nhiều làn sóng đổi mới rồi suy thoái.
Sự bùng nổ gần đây của mạng nơ-ron sâu (DNN) — tức là mạng có
nhiều lớp phi tuyến hơn so với các
mô hình trước đây — có lẽ bắt nguồn từ sự gia tăng
năng lực tính toán sẵn có và sự sẵn có của các tập dữ liệu lớn. Điều này đã thúc đẩy
một loạt đổi mới tăng trưởng theo mọi hướng, và trong nhiều
bài toán mô hình hóa, mạng nơ-ron sâu hiện nay cho kết quả tốt hơn hẳn
so với bất kỳ phương pháp cạnh tranh nào.

Mặc dù đạt được nhiều thành công, việc ứng dụng DNN trong xử lý tiếng nói
không tránh khỏi những vấn đề nhất định. Ví dụ,

-   Việc huấn luyện DNN cho một nhiệm vụ cụ thể trên một tập dữ liệu nhất định thường
    không giúp chúng ta hiểu sâu hơn về bài toán. Đó là một hộp
    đen. Làm sao chúng ta biết được mô hình có đáng tin cậy hay không? Một mô hình
    nhận dạng tiếng nói được huấn luyện trên ngôn ngữ A không giúp ích nhiều cho việc
    nhận dạng tiếng nói ngôn ngữ B (mặc dù quá trình thiết kế bộ
    nhận dạng giúp chúng ta hiểu thêm về các ngôn ngữ).
-   Việc huấn luyện DNN rất nhạy cảm với dữ liệu được sử dụng.
    Các mô hình có thể bị ảnh hưởng bởi những thiên kiến ẩn, khiến
    hiệu suất giảm đối với các nhóm
    người thiếu đại diện cụ thể.
-   Một DNN đã huấn luyện là "một giải pháp" cho một bài toán cụ thể, nhưng nó
    không trực tiếp cho ta biết giải pháp đó tốt đến mức nào.
    Ví dụ, nếu một tập dữ liệu biểu diễn một đường tròn trong mặt phẳng 2D, thì
    có thể mô hình hóa chính xác tập dữ liệu đó bằng một mạng
    nơ-ron với các hàm phi tuyến là sigmoid. Mạng nơ-ron
    chỉ cần đủ lớn và nó có thể hoàn thành nhiệm vụ. Tuy nhiên, khi đó
    mạng phức tạp hơn nhiều bậc so với
    phương trình đường tròn. Nghĩa là, mặc dù việc huấn luyện mạng
    đã thành công và mô hình tương đối chính xác, nó không cho ta
    biết liệu độ phức tạp của mạng có cùng mức với
    độ phức tạp của bài toán hay không.

Để giải quyết những vấn đề này, một xu hướng gần đây trong thiết kế mô hình là
quay trở lại các mô hình thiết kế cổ điển, trong đó các mô hình dựa trên sự hiểu biết thấu đáo
về bài toán. Các tham số của những mô hình này sau đó
được huấn luyện bằng các phương pháp học máy.

## Cấu trúc mạng

Về nguyên tắc, mạng nơ-ron được xây dựng từ các khối xây dựng đơn giản,
trong đó loại khối xây dựng phổ biến nhất là

$$ y = f(A x + b) $$

trong đó *x* là vectơ đầu vào, ma trận *A* và vectơ *b* là các hằng số,
*f* là một hàm phi tuyến chẳng hạn như sigmoid theo từng phần tử, và *y*
là đầu ra. Đây thường được gọi là một *lớp*. Các lớp có thể
được xếp chồng liên tiếp sao cho, ví dụ, một mạng ba lớp
sẽ là

$$ \begin{matrix} y_1 &= f(A_1 x + b_1)\\ y_2 &= f(A_2 y_1 +
b_3)\\ y_{out} &= f(A_3 y_2 + b_3). \end{matrix} $$

### Mạng nơ-ron sâu (DNN)

Mạng nơ-ron sâu (DNN) là một mạng nơ-ron nhân tạo (ANN) với
nhiều lớp giữa lớp đầu vào và lớp đầu ra. Nhiều chuyên gia định nghĩa
mạng nơ-ron sâu là mạng có một lớp đầu vào, một lớp đầu ra
và ít nhất một lớp ẩn ở giữa. Mỗi lớp thực hiện
các loại phân loại và sắp xếp cụ thể trong một quá trình mà một số người gọi
là "phân cấp đặc trưng". Một trong những ứng dụng chính của các mạng
nơ-ron tinh vi này là xử lý dữ liệu chưa gán nhãn hoặc phi cấu trúc. Thuật ngữ
"học sâu" cũng được dùng để chỉ các mạng nơ-ron sâu này,
vì học sâu đại diện cho một dạng cụ thể của học
máy, trong đó các công nghệ sử dụng các khía cạnh của trí tuệ nhân tạo
tìm cách phân loại và sắp xếp thông tin theo những cách vượt xa các
giao thức đầu vào/đầu ra đơn giản.

### Cơ bản về mạng nơ-ron

Nơ-ron: Đây là thành phần cấu trúc cơ bản của mạng nơ-ron. Khi nhận được
thông tin, chúng ta xử lý nó và sau đó tạo ra đầu ra.
Tương tự, một nơ-ron nhận đầu vào, xử lý nó và tạo ra
đầu ra, sau đó được gửi đến các nơ-ron khác để tiếp tục xử lý hoặc
là đầu ra cuối cùng.

Trọng số (Weights): Khi một đầu vào đi vào nơ-ron, nó được nhân với một trọng số.
Ban đầu, các trọng số được khởi tạo và chúng được cập nhật trong
quá trình huấn luyện mô hình. Khi quá trình huấn luyện kết thúc, mạng nơ-ron
gán giá trị trọng số cao hơn cho các đầu vào mà nó cho là quan trọng
hơn so với những đầu vào được coi là ít quan trọng hơn. 

  

Độ chệch (Bias): Ngoài các trọng số, một thành phần tuyến tính khác được áp dụng lên
đầu vào, gọi là độ chệch. Nó được cộng vào kết quả của phép nhân trọng số
với đầu vào. Về cơ bản, độ chệch được thêm vào để thay đổi
phạm vi của đầu vào đã nhân với trọng số.

### Hàm kích hoạt:

1.  Sigmoid: Hàm này cho phép giảm các giá trị cực đoan hoặc bất thường trong
    dữ liệu hợp lệ mà không loại bỏ chúng: nó chuyển đổi các biến
    độc lập có phạm vi gần như vô hạn thành các xác suất đơn giản trong khoảng
    từ 0 đến 1. Phần lớn đầu ra của nó sẽ rất gần với các cực trị 0
    hoặc 1.
    $$ sigmoid(x) = \frac{1}{1+e^{-x}} $$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-7,7,100)
plt.plot(x, 1./(1. + np.exp(-x)))
plt.xlabel('$x$')
plt.ylabel('$sigmoid(x)$')
plt.show()
```

      

2.  ReLU (Rectified Linear Unit): Hàm này có đầu ra bằng 0 nếu đầu vào nhỏ
    hơn 0, và đầu ra bằng chính đầu vào trong trường hợp ngược lại. Tức là, nếu đầu vào lớn
    hơn 0, đầu ra bằng chính đầu vào. Hoạt động của ReLU
    gần với cách các nơ-ron sinh học của chúng ta hoạt động hơn.
    $$ ReLU(x) = \max(x,0) = \begin{cases} x, & x>0\\ 0, & \text{otherwise}.\end{cases} $$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-7,7,100)
plt.plot(x, (x>0)*x)
plt.xlabel('$x$')
plt.ylabel('$ReLU(x)$')
plt.show()
```

      

3.  [Softmax](https://en.wikipedia.org/wiki/Softmax_function): Một biến thể
    của hàm $\max$ thông thường trên một vectơ, sao cho nó có đạo hàm liên tục
    tại mọi nơi. Nghĩa là, nó ánh xạ
    đầu ra vào khoảng $[0,1]$ và đồng thời đảm bảo
    rằng tổng bằng 1. Do đó, đầu ra của softmax là
    một phân phối xác suất.
    $$  SoftMax (\mathbf {x_k} )=\frac {e^{x_k}}{\sum _{j=1}^{K}e^{x_j}}.$$
    

5. Lớp đầu vào / Lớp đầu ra / Lớp ẩn: Lớp đầu vào nhận dữ liệu đầu vào
và là lớp đầu tiên của mạng. Lớp đầu ra là lớp
tạo ra kết quả và là lớp cuối cùng của mạng. Các
lớp xử lý là các lớp ẩn bên trong mạng. Những lớp ẩn
này thực hiện các nhiệm vụ cụ thể trên dữ liệu đầu vào
và chuyển đầu ra do chúng tạo ra đến lớp tiếp theo. Lớp đầu vào
và lớp đầu ra là những lớp mà chúng ta có thể nhìn thấy, trong khi các
lớp trung gian là lớp ẩn.

6. MLP (Multi Layer Perceptron): Trong mạng đơn giản nhất, chúng ta có
một lớp đầu vào, một lớp ẩn và một lớp đầu ra. Mỗi lớp có
nhiều nơ-ron và tất cả các nơ-ron trong mỗi lớp được kết nối với tất cả
các nơ-ron ở lớp tiếp theo. Những mạng này còn có thể được gọi là
mạng kết nối đầy đủ.

7. Hàm chi phí hoặc hàm mất mát: Khi huấn luyện một mạng, mục tiêu chính là
dự đoán đầu ra càng gần với giá trị thực càng tốt. Do đó,
hàm chi phí/mất mát được sử dụng để đo lường độ chính xác này. Hàm chi phí hoặc mất mát
trừng phạt mạng khi nó mắc lỗi. Mục tiêu chính
khi vận hành mạng là tăng độ chính xác dự đoán và
giảm lỗi, từ đó tối thiểu hóa hàm chi phí.

8. Hạ gradient (Gradient Descent): Đây là một thuật toán tối ưu hóa được sử dụng để tối thiểu hóa
một hàm nào đó bằng cách lặp đi lặp lại di chuyển theo hướng giảm nhanh nhất
được xác định bởi gradient âm.

9. Tốc độ học (Learning Rate): Tốc độ học là một siêu tham số kiểm soát
mức độ thay đổi mô hình để phản ứng với lỗi ước tính mỗi
khi các trọng số mô hình được cập nhật. Việc chọn tốc độ học
là một thách thức vì giá trị quá nhỏ có thể dẫn đến quá trình huấn luyện kéo dài
và có thể bị mắc kẹt, trong khi giá trị quá lớn có thể dẫn đến việc học được
một tập trọng số dưới tối ưu quá nhanh hoặc quá trình huấn luyện không ổn định.
Tốc độ học có thể là siêu tham số quan trọng nhất khi cấu hình
mạng nơ-ron. Do đó, điều quan trọng là phải biết cách khảo sát
ảnh hưởng của tốc độ học lên hiệu suất mô hình và xây dựng
trực giác về động lực học của tốc độ học lên hành vi mô hình.

10. Lan truyền ngược (Backpropagation): Khi định nghĩa một mạng nơ-ron, chúng ta gán các giá trị
trọng số và độ chệch ngẫu nhiên cho các nút. Sau khi nhận được đầu ra
cho một vòng lặp duy nhất, chúng ta có thể tính toán lỗi của mạng. Lỗi
này sau đó được truyền ngược lại mạng cùng với gradient của
hàm chi phí để cập nhật các trọng số của mạng. Các trọng số này
sau đó được cập nhật sao cho lỗi trong các vòng lặp tiếp theo được giảm thiểu.
Việc cập nhật các trọng số sử dụng gradient của hàm chi phí
được gọi là lan truyền ngược. Trong lan truyền ngược, sự di chuyển của
mạng diễn ra theo chiều ngược, lỗi cùng với gradient chảy ngược từ
lớp đầu ra qua các lớp ẩn và các trọng số được cập nhật.

11. Lô (Batches): Khi huấn luyện mạng nơ-ron, thay vì gửi
toàn bộ đầu vào một lần, chúng ta chia đầu vào thành nhiều phần
có kích thước bằng nhau một cách ngẫu nhiên. Huấn luyện dữ liệu theo từng lô làm cho mô hình
tổng quát hơn so với mô hình được xây dựng khi toàn bộ tập dữ liệu
được đưa vào mạng một lần.

12. Epoch: Một epoch là một vòng lặp huấn luyện duy nhất của tất cả các lô
trong cả lan truyền xuôi và lan truyền ngược. Như vậy, 1 epoch là một lần truyền xuôi
và ngược của toàn bộ dữ liệu đầu vào. Mặc dù rất có khả năng
nhiều epoch hơn sẽ cho thấy độ chính xác cao hơn của mạng, nhưng
cũng sẽ mất nhiều thời gian hơn để mạng hội tụ. Cũng cần lưu
ý rằng nếu số epoch quá cao, mạng
có thể bị quá khớp.

13. Dropout: Đây là một kỹ thuật chính quy hóa để ngăn chặn hiện tượng quá khớp
của mạng. Trong quá trình huấn luyện, một số nơ-ron trong lớp ẩn
bị loại bỏ ngẫu nhiên. 

14. Chuẩn hóa lô (Batch Normalization): Kỹ thuật này chuẩn hóa lớp đầu vào bằng cách điều chỉnh và
chia tỷ lệ các hàm kích hoạt. Ví dụ, nếu chúng ta có các đặc trưng từ 0 đến 1
và một số từ 1 đến 1000, chúng ta nên chuẩn hóa chúng để tăng tốc quá trình học.

### Mạng nơ-ron tích chập

CNN lần đầu tiên được đề xuất bởi [1]. Nó đã được áp dụng thành công bởi
[2] cho bài toán phân loại chữ số viết tay. Hiện nay, đây là
mô hình mạng nơ-ron phổ biến nhất được sử dụng cho bài toán phân loại
ảnh. Ưu điểm của CNN so với DNN bao gồm: CNN được tối ưu hóa cao
cho việc xử lý ảnh 2D và 3D, và có khả năng học
và trích xuất các đặc trưng trừu tượng 2D một cách hiệu quả. Ngoài ra, CNN có
số lượng tham số ít hơn đáng kể so với mạng kết nối đầy đủ có cùng
kích thước.

Hình 4.1. thể hiện kiến trúc tổng thể của CNN. CNN chủ yếu bao gồm
hai phần: bộ trích xuất đặc trưng và bộ phân loại. Trong mô-đun
trích xuất đặc trưng, mỗi lớp của mạng nhận đầu ra từ
lớp trước đó ngay trước nó làm đầu vào và chuyển đầu ra của nó làm
đầu vào cho lớp tiếp theo. Kiến trúc CNN chủ yếu bao gồm ba
loại lớp: tích chập, gộp và phân loại.

Hình 4.1. Kiến trúc tổng thể của mạng nơ-ron tích chập.

Các lớp chính trong mạng nơ-ron tích chập bao gồm:

-   Lớp tích chập: Một "bộ lọc" quét qua ảnh, quét
    một vài điểm ảnh tại một thời điểm và tạo ra bản đồ đặc trưng dự đoán
    lớp mà mỗi đặc trưng thuộc về. Như vậy, trong lớp này, các bản đồ
    đặc trưng từ các lớp trước được tích chập với các hạt nhân có thể học. Đầu ra
    của các hạt nhân đi qua một hàm kích hoạt tuyến tính hoặc phi tuyến
, chẳng hạn như sigmoid, hyperbolic tangent, softmax, rectified
    linear và hàm đồng nhất) để tạo thành các bản đồ đặc trưng đầu ra.
    Mỗi bản đồ đặc trưng đầu ra có thể được kết hợp với nhiều hơn một
    bản đồ đặc trưng đầu vào.

-   Lớp gộp (Pooling): Lớp này giảm lượng thông tin trong mỗi đặc trưng
    thu được ở lớp tích chập trong khi giữ lại thông tin
    quan trọng nhất (thường có nhiều vòng
    tích chập và gộp liên tiếp).

-   Lớp kết nối đầy đủ đầu vào (làm phẳng): Lớp này nhận đầu ra của
    các lớp trước, "làm phẳng" chúng và chuyển thành một vectơ đơn
    có thể làm đầu vào cho giai đoạn tiếp theo.

-   Lớp kết nối đầy đủ thứ nhất: Lớp này nhận các đầu vào từ
    phần phân tích đặc trưng và áp dụng các trọng số để dự đoán nhãn chính xác. 

-   Lớp kết nối đầy đủ đầu ra: Lớp này đưa ra xác suất cuối cùng cho
    mỗi nhãn.

  

Các đặc trưng cấp cao được suy ra từ các đặc trưng được truyền từ các
lớp cấp thấp hơn. Khi các đặc trưng truyền đến lớp hoặc cấp cao nhất,
kích thước của các đặc trưng bị giảm dựa trên kích thước của hạt nhân
cho các phép toán tích chập và gộp tương ứng. Tuy nhiên,
số lượng bản đồ đặc trưng thường tăng lên để biểu diễn các đặc trưng tốt hơn
của ảnh đầu vào nhằm đảm bảo độ chính xác phân loại. Đầu ra của
lớp cuối cùng của CNN được sử dụng làm đầu vào cho một mạng
kết nối đầy đủ. Mạng nơ-ron truyền thẳng được sử dụng làm lớp
phân loại vì chúng cho hiệu suất tốt hơn.

Hình 4.2 Các kiến trúc CNN phổ biến. Hình được lấy từ
https://www.aismartz.com/blog/cnn-architectures/.

Ứng dụng của CNN:

-   Xử lý ảnh và Thị giác máy tính: CNN đã được áp dụng thành công
    trong nhiều bài toán phân loại ảnh khác nhau [7–11].

-   Xử lý tiếng nói: CNN đã được áp dụng thành công trong nhiều
    ứng dụng xử lý tiếng nói khác nhau như tăng cường tiếng nói [8] và
    gán nhãn âm thanh [9].

-   Hình ảnh y tế: CNN cũng đã được sử dụng rộng rãi trong nhiều
    lĩnh vực xử lý hình ảnh y tế bao gồm các bài toán phân loại, phát hiện và
    phân đoạn [10].

Kỹ thuật huấn luyện

1.  Lớp lấy mẫu con hoặc Lớp gộp: Hai kỹ thuật khác nhau đã
    được sử dụng để triển khai các mạng sâu trong
    lớp lấy mẫu con hoặc lớp gộp: trung bình và gộp cực đại. Trong khi
    gộp trung bình tính giá trị trung bình cho mỗi vùng trên
    bản đồ đặc trưng, thì gộp cực đại tính giá trị lớn nhất cho mỗi
    vùng của bản đồ đặc trưng. 

2.  Đệm (Padding): Kỹ thuật này thêm một lớp số 0 bổ sung xung quanh ảnh để
    ảnh đầu ra có cùng kích thước với đầu vào. 

3.  Tăng cường dữ liệu (Data Augmentation): Đây là việc bổ sung dữ liệu mới được suy ra từ
    dữ liệu đã cho. Điều này có thể có lợi cho việc dự đoán. Nó
    bao gồm xoay, cắt xiên, phóng to, cắt xén, lật và
    thay đổi mức độ sáng.

      

Lưu ý rằng hiệu suất của CNN phụ thuộc nhiều vào nhiều
siêu tham số: số lượng lớp, số lượng bản đồ đặc trưng trong mỗi lớp,
việc sử dụng dropout, chuẩn hóa lô, v.v. Do đó, điều quan trọng là
bạn nên tinh chỉnh các siêu tham số của mô hình trước bằng cách tiến hành nhiều
thử nghiệm. Sau khi tìm được bộ siêu tham số phù hợp, bạn cần
huấn luyện mô hình trong nhiều epoch.

  

### Mạng nơ-ron hồi quy (RNN)

Các chủ đề sẽ được trình bày:

-   Cơ bản về RNN
-   Vấn đề tiêu biến và bùng nổ gradient
-   Bộ nhớ dài ngắn hạn (LSTM) 

**Cơ bản về RNN**

Các mạng nơ-ron truyền thẳng tiêu chuẩn (tức là DNN và CNN) là
các bộ sinh hàm liên kết đầu ra phù hợp với đầu vào. Tuy nhiên,
một số loại dữ liệu có tính chất tuần tự. Mạng nơ-ron
hồi quy (RNN) xử lý các chuỗi như giá cổ phiếu, câu văn từng phần tử một
trong khi duy trì một bộ nhớ (gọi là trạng thái) về những gì đã xảy ra
trước đó trong chuỗi. Hồi quy có nghĩa là đầu ra tại bước thời gian
hiện tại trở thành đầu vào cho bước thời gian tiếp theo. Tại mỗi phần tử của
chuỗi, mô hình xem xét không chỉ đầu vào hiện tại mà còn những gì
nó nhớ về các phần tử trước đó. Bộ nhớ này cho phép
mạng học các phụ thuộc dài hạn trong một chuỗi, nghĩa là nó có thể
xem xét toàn bộ ngữ cảnh khi đưa ra dự đoán chẳng hạn như
dự đoán từ tiếp theo trong một câu. RNN được thiết kế để mô phỏng
cách con người xử lý chuỗi: chúng ta xem xét toàn bộ câu khi
hình thành câu trả lời thay vì từng từ riêng lẻ. Ví dụ, hãy xem xét
câu sau:

"Buổi hòa nhạc thật nhàm chán trong vài phút đầu nhưng sau đó lại vô cùng
thú vị."

Một mô hình học máy xem xét các từ một cách riêng lẻ có lẽ
sẽ kết luận câu này mang nghĩa tiêu cực. Ngược lại, một RNN
có thể nhận ra các từ "nhưng" và "vô cùng thú vị" và nhận ra rằng
câu chuyển từ tiêu cực sang tích cực vì nó đã xem xét
toàn bộ chuỗi. Đọc toàn bộ chuỗi cho chúng ta ngữ cảnh để
xử lý ý nghĩa của nó, một khái niệm được mã hóa trong mạng nơ-ron hồi quy.

  

  

Hình 4.3 Dạng tổng quát của RNN

  

  

Phần bên trái của Hình 4.3 cho thấy mối quan hệ đầu vào-đầu ra của một
mạng nơ-ron tiêu chuẩn được thay đổi sao cho đầu ra được đưa vào
đầu vào. Nhưng phần bên phải của Hình 4.3 cho thấy sơ đồ được mở ra
theo thời gian. Đầu vào là dữ liệu tuần tự $ (x_1,........,x_T) $
và đầu ra là $ (o_1,........,o_T) $ . Đầu ra của một mạng
nơ-ron được đưa vào mạng nơ-ron thành phần tiếp theo ở
giai đoạn sau như một phần của đầu vào. Vì vậy đầu ra $ o_t $ phụ thuộc vào tất cả
các đầu vào $ (x_1,........,x_T) $ . Có thể xảy ra trường hợp
không phải là ý hay khi làm cho đầu ra $ o_1 $

chỉ phụ thuộc vào $ x_1 $ . Thực tế $ o_1 $ bản thân nó nên được
tạo ra trong ngữ cảnh.

RNN có thể được sử dụng trong nhiều ứng dụng khác nhau như dịch máy,
nhận dạng tiếng nói, tạo mô tả ảnh, gán nhãn video và
mô hình hóa ngôn ngữ. 

Ưu điểm của mạng nơ-ron hồi quy

1.  RNN ghi nhớ mọi thông tin qua thời gian. Nó
    hữu ích trong dự đoán chuỗi thời gian chính nhờ khả năng
    ghi nhớ các đầu vào trước đó. Đây được gọi là Bộ nhớ Dài Ngắn hạn.
2.  Mạng nơ-ron hồi quy thậm chí còn được sử dụng cùng với các lớp tích chập để
    mở rộng vùng lân cận điểm ảnh hiệu quả.

Nhược điểm của mạng nơ-ron hồi quy

1.  Vấn đề tiêu biến và bùng nổ gradient.
2.  Huấn luyện RNN là một nhiệm vụ rất khó khăn.
3.  Không thể xử lý các chuỗi rất dài nếu sử dụng tanh hoặc relu làm
    hàm kích hoạt.

**Vấn đề tiêu biến và bùng nổ gradient**

RNN rất khó huấn luyện. Hãy xem xét lý do. Biểu thức $ \frac{\partial E}{\partial h^l} $ được sử dụng trong thuật toán lan truyền ngược là một
tích của một chuỗi dài các ma trận: $ \left( \frac {\partial
E}{\partial h^l} \right ) = \left ( \frac {\partial
z^{l+1}}{\partial h^l} \right ) \left ( \frac {\partial
h^{l+1}}{\partial z^{l+1}} \right )...\left ( \frac {\partial
h^l}{\partial z^L} \right ) \left ( \frac {\partial E}{\partial
h^L} \right ), $ Với đầu vào $ x=h^0 $ , chuỗi này là
dài nhất. Nếu hàm kích hoạt $ sigmoid(t) $ là hàm sigmoid
$ sigmoid(t) = \frac{1}{1+e^{-t}} $ thì đạo hàm của nó là $
sigmoid'(t) = \frac{e^t} { \left( 1 + e^t \right ) ^2 } $ rất
nhỏ đến mức nó thực tế biến mất ngoại trừ tại một khoảng nhỏ
gần 0. Đây là một trong những lý do tại sao hàm kích hoạt ReLU
được ưu tiên. Nhưng nó không phải là giải pháp triệt để, vì các giá trị đầu vào âm
cũng làm tiêu biến gradient và giữ nó ở đó. Ngay cả khi tránh được
vấn đề tiêu biến gradient hoàn toàn, phép nhân ma trận dài
nói chung có thể làm gradient tiêu biến hoặc bùng nổ. Vấn đề này
càng trở nên nghiêm trọng hơn trong trường hợp của RNN, vì RNN thường
yêu cầu chuỗi lan truyền ngược dài không chỉ qua các lớp của
mạng nơ-ron các tế bào thành phần mà còn qua các
tế bào khác nhau. Do đó, RNN rất khó huấn luyện.


**Bộ nhớ dài ngắn hạn (LSTM)**

Bộ nhớ Dài Ngắn hạn (LSTM) lần đầu tiên được đề xuất bởi Hochreiter và
Schmidhuber [10] như một giải pháp cho vấn đề tiêu biến gradient. Nhưng
nó đã không thu hút được nhiều sự chú ý cho đến khi mọi người nhận ra rằng nó thực sự
cung cấp một giải pháp tốt cho vấn đề tiêu biến và bùng nổ gradient
của RNN như đã mô tả ở trên. Chúng ta sẽ chỉ mô tả kiến trúc của
tế bào của nó. Có nhiều biến thể trong kiến trúc tế bào, nhưng chúng tôi chỉ trình bày
những kiến thức cơ bản. 

Trung tâm của RNN là một lớp được tạo thành từ các tế bào bộ nhớ. Tế bào
phổ biến nhất hiện nay là Bộ nhớ Dài Ngắn hạn (LSTM), duy trì
một trạng thái tế bào cũng như một bộ nhớ mang để đảm bảo rằng tín hiệu
(thông tin dưới dạng gradient) không bị mất khi chuỗi được
xử lý. Tại mỗi bước thời gian, LSTM xem xét từ hiện tại,
bộ nhớ mang và trạng thái tế bào. LSTM có 3 cổng khác nhau và các vectơ
trọng số: có một cổng "quên" để loại bỏ thông tin không liên quan;
một cổng "đầu vào" để xử lý đầu vào hiện tại và một cổng "đầu ra" để
tạo dự đoán tại mỗi bước thời gian. Tuy nhiên, như Chollet đã chỉ ra,
việc cố gắng gán ý nghĩa cụ thể cho từng
thành phần trong tế bào là không có kết quả.

Hình 4.4. Cấu trúc tế bào LSTM

Cấu trúc của LSTM được thể hiện trong Hình 4.4. Vectơ đầu vào là $
x_t $ . Trạng thái tế bào được ký hiệu bởi $ c_{t-1} $ và trạng thái ẩn
$ h_{t-1} $ được đưa vào tế bào LSTM và $ c_t $ và 
$ h_t $ được đưa vào tế bào tiếp theo. Bên trong, nó có bốn trạng thái: 
$ i_t $ (đầu vào), $ f_t $ (quên), $ o_t $ (đầu ra), và 
$ g_t $ . Trạng thái quên $ f_t $ được tính như đầu ra sigmoid
của một mạng với $ x_t $ và $ h_{t-1} $ được đưa vào
làm đầu vào. Vì $ f_t $ là đầu ra sigmoid, mỗi phần tử trong nó
có giá trị từ 0 đến 1. Nếu gần bằng 0, nó xóa $
c_{t-1} $ bằng phép nhân; nếu gần bằng 1, nó giữ $
c_{t-1} $ bằng phép nhân. Do đó, $ f_t $ được gọi là
"trạng thái quên" vì tính chất này. Trạng thái đầu vào $ i_t $
và trạng thái đầu ra $ o_t $ được tính bằng cách sử dụng (1) và (3)
tương ứng. Trạng thái $ g_t $ cũng được tính tương tự ngoại trừ
nó là đầu ra dạng tanh. Điều này cho dấu $ \pm $
. Tích $ i_t $ $ \otimes $ $ g_t $ sau đó được cộng vào $ c_t $
, và điều này mang lại thông tin mới cho trạng thái tế bào. Trạng thái ẩn $ h_t $
được tính như trong (6), và đầu ra tế bào $ y_t $ giống như
$ h_t $ . Toàn bộ sơ đồ được mô tả trong Hình 4.4. 

$$ 
\begin{array}{lcll}
i_t (input) &=& \sigma (W_{ix}x_t + W_{ih}h_{t-1} + b_i) &(1) \\
f_t (forget) &=& \sigma (W_{fx}x_t + W_{fh}h_{t-1} + b_f) & (2) \\
o_t (output)&=& \sigma (W_{ox}x_t + W_{oh}h_{t-1} + b_o) & (3) \\
g_t &=& tanh (W_{gx}x_t + W_{gh}h_{t-1} + b_g) & (4) \\
c_t (cell\ state) &=& ft\otimes c_{t-1} + i_t \otimes g_t & (5)\\
h_t (hidden\ state) &=& o_t \otimes tanh(c_t) &(6)\\
y_t (cell\ output) &=& h_t & (7)
\end{array}
$$
