# An ninh và quyền riêng tư trong công nghệ tiếng nói (Security and privacy in speech technology)


***TUYÊN BỐ MIỄN TRỪ TRÁCH NHIỆM:*** *Tài liệu này được biên soạn nhằm mục đích giới thiệu các câu hỏi về an ninh và quyền riêng tư trong công nghệ tiếng nói cho sinh viên ngành kỹ thuật, giúp họ hiểu được những vấn đề cốt lõi. Tài liệu này không phải là một văn bản pháp lý. Trong các ứng dụng công nghệ và thu thập dữ liệu thực tế, bạn phải tham khảo ý kiến của các chuyên gia pháp lý để xác định xem mình có tuân thủ pháp luật hay không. Bạn tự chịu trách nhiệm về hành vi của mình.*

[Quyền riêng tư](https://en.wikipedia.org/wiki/Right_to_privacy) (right to privacy) là một khái niệm được chấp nhận rộng rãi mặc dù định nghĩa của nó có thể thay đổi tùy ngữ cảnh. Tuy nhiên, có một điều rõ ràng là mọi người thường có xu hướng nghĩ rằng một số thứ là "của họ", nghĩa là họ có quyền sở hữu đối với các *vật chất* hoặc *phi vật chất*, bao gồm cả thông tin về bản thân họ. Do đó, một định nghĩa khả dĩ cho quyền riêng tư là "sự không bị người khác chú ý" và tương ứng, an ninh (security) có thể được định nghĩa là sự bảo vệ đối với những gì một người sở hữu, bao gồm cả những thứ hữu hình và vô hình. Tuy nhiên, cần nhấn mạnh rằng không có các định nghĩa chung được chia sẻ và chấp nhận rộng rãi; đặc biệt, cộng đồng pháp lý có rất nhiều định nghĩa khác nhau tùy thuộc vào bối cảnh và lĩnh vực áp dụng.

Dưới góc nhìn của công nghệ tiếng nói, an ninh và quyền riêng tư có hai khía cạnh chính:
-   An ninh và quyền riêng tư của dữ liệu liên quan đến tín hiệu tiếng nói và
-   Sự bảo vệ chống lại các cuộc tấn công sử dụng tín hiệu tiếng nói làm công cụ.

Khía cạnh thứ hai chủ yếu liên quan đến danh tính người nói; ví dụ, những kẻ lừa đảo có thể tổng hợp tiếng nói bắt chước (giả mạo - spoof) một người mục tiêu để truy cập trái phép vào các hệ thống bị hạn chế, chẳng hạn như tài khoản ngân hàng của người đó. Các kịch bản sử dụng như vậy chủ yếu thuộc về chủ đề thảo luận của phần [nhận dạng và xác thực người nói](Recognition/Speaker_Recognition_and_Verification.md), và sẽ không được thảo luận thêm ở đây.

Cần lưu ý rằng trong phân khúc độc lập của dịch vụ thoại (kết nối điện thoại truyền thống), quyền riêng tư và an ninh đã có các tiêu chuẩn đạo đức cũng như luật pháp được thiết lập rất chặt chẽ. Ở hầu hết các quốc gia, các cuộc gọi điện thoại là riêng tư theo nghĩa chỉ những người tham gia cuộc gọi "dự kiến" mới được phép lắng nghe và thậm chí việc ghi âm cuộc gọi đôi khi cũng bị hạn chế. Việc nghe lén bí mật thường chỉ được phép đối với lực lượng cảnh sát và ngay cả với họ, điều đó cũng chỉ được phép trong các tình huống được quy định đặc biệt, chẳng hạn như có sự cho phép của tòa án hoặc thẩm phán.

{cite}`edps2019techdispatch,nautsch2019gdpr` 


## Các mô hình hệ thống
### Tương tác hoàn toàn giữa con người với con người
Tiếng nói là công cụ giao tiếp, vì vậy thông thường chúng ta luôn thảo luận về các tương tác giữa hai chủ thể, ví dụ như Alice và Bob. Sự tương tác giữa họ là chức năng mong muốn, do đó thông tin trao đổi ở đó được ngầm định là được phép một cách rõ ràng. Bằng cách chọn nói chuyện với nhau, cả hai đều tiết lộ thông tin cá nhân ở mức độ mà tiếng nói của họ chứa đựng thông tin đó.

#### Tương tác thứ cấp (Secondary interactions)

Một câu hỏi cấp hai là về bên thứ ba (Eve), những người không tham gia trực tiếp vào cuộc giao tiếp tiếng nói chính. Câu hỏi đặt ra là bên thứ ba được phép tham gia vào cuộc tương tác ở mức độ nào. Như một ví dụ thực tế, giả sử Alice và Bob đang có một bữa tối lãng mạn tại một nhà hàng. Người phục vụ bàn Eve được phép tiếp cận cuộc thảo luận của Alice và Bob ở mức độ nào? Rõ ràng Eve có một số nhiệm vụ bắt buộc phải thực hiện khiến việc tiếp xúc là không thể tránh khỏi. Liệu Alice và Bob có tạm dừng cuộc thảo luận của họ khi Eve đến gần hay không?


````python
%%itikz --temp-dir 
\documentclass{standalone}
\usepackage[utf8]{inputenc}
\usepackage{tikz}
\usepackage{verbatim}


\usepackage{pgfplots}
\DeclareUnicodeCharacter{2212}{−}
\usepgfplotslibrary{groupplots,dateplot}
\usetikzlibrary{patterns,shapes.arrows}
\usetikzlibrary {fit} 
\usetikzlibrary{shapes.geometric,positioning}
\usetikzlibrary{bending}
\pgfplotsset{compat=newest}

\begin{document}


\begin{tikzpicture}
  \node[ellipse,draw,minimum width=40pt] at (0,0) (alice) {Alice};
  \node[ellipse,draw,minimum width=40pt] at (2,0)  (bob) {Bob};
  \draw[->, line width=1pt] (alice) to [bend left] (bob);
  \draw[->, line width=1pt] (bob) to [bend left] (alice);
  %\node[fit=(alice) (bob), inner sep=1pt] (p) {};% {\parbox{2.4cm}{\centering Desired speech interaction\vspace{-1cm}}};
  %\node[draw=gray,inner sep=6pt,very thick,ellipse,fit=(alice) (bob),label={[label distance=.01cm]180:{\parbox{1.7cm}{\centering Desired\\ interaction}}}] (private) {};
  %\node[draw=gray,inner sep=6pt,very thick,ellipse,fit=(alice) (bob) (p)] (private) {};
  \node[draw=gray,very thick,ellipse,minimum width=115pt, minimum height = 40pt] at (1,0) (private) {};

  \node[ellipse,draw] at (1,2.5) (eve) {Eve};
  \draw[color=purple,->, line width=1pt] (eve) to [bend left] (private);
  \draw[color=purple,->, line width=1pt] (private) to [bend left] (eve);
  
  \node at (-2,0) {\parbox{2cm}{\centering Private speech interaction}};
  \node at (-2,2.5) {\parbox{2cm}{\centering Restricted interactor}};
\end{tikzpicture}

\end{document}
````

Lưu ý rằng ở đây chúng tôi đã gắn nhãn Eve là người tương tác "*bị hạn chế*" (restricted) chứ không phải người tương tác "*trái phép*" (unauthorized). Nếu quyền truy cập là trái phép, rõ ràng là Eve không được phép tiếp cận cuộc hội thoại tiếng nói, điều này thường rất đơn giản để xử lý. Ngược lại, từ "bị hạn chế" ngụ ý rằng quyền truy cập hoàn toàn không được cấp, nhưng một phần quyền truy cập vẫn có thể được cho phép. Do đó, câu hỏi không phải là "*có nên*" cấp quyền truy cập hay không mà là "*cấp bao nhiêu?*".

#### Quyền sở hữu và sự riêng tư cá nhân

Quyền riêng tư có mối liên hệ chặt chẽ với quyền sở hữu tài sản phi vật chất, cụ thể là thông tin. Quyền sở hữu như vậy cũng có thể được chuyển dịch thành câu hỏi: *ai có quyền kiểm soát* đối với một thông tin nào đó? Dưới góc độ *riêng tư cá nhân* (personal privacy), rõ ràng là nó liên quan đến thông tin mà một người duy nhất có quyền tuyên bố sở hữu.

Tuy nhiên, tiếng nói lại phức tạp hơn. Tiếng nói là một hình thức giao tiếp và do đó liên quan đến tương tác giữa hai bên. Các cuộc hội thoại thường có thể dẫn đến việc đồng kiến tạo ý nghĩa (co-creation of meaning), nơi thông tin mới được tạo ra thông qua cuộc đối thoại dưới dạng thức mà không bên nào có thể tự mình tạo ra đơn độc {cite:p}`gasiorek2018message`. Do đó, không ai trong số các bên tham gia có thể tuyên bố sở hữu duy nhất đối với thông tin đó, mà quyền sở hữu là chia sẻ. Hiện nay, chúng ta chưa có các công cụ để xử lý quyền sở hữu chia sẻ như vậy.

Chúng ta có thể coi thiết bị đó như một thiết bị cục bộ thuần túy. Nếu không có ai khác có quyền truy cập vào thiết bị đó, thì tất cả thông tin vẫn nằm trong tầm kiểm soát trực tiếp của người dùng. Ngay cả khi thiết bị kết nối với dịch vụ đám mây từ xa, nếu thông tin vẫn được giữ nghiêm ngặt trong phạm vi dịch vụ mong muốn, thì có rất ít vấn đề cần phải cân nhắc.

## Thông tin chứa trong tín hiệu tiếng nói

Tín hiệu tiếng nói chứa đựng rất nhiều thông tin đa dạng, trong đó có nhiều thông tin có khả năng mang tính riêng tư hoặc thậm chí là nhạy cảm. Việc liệt kê đầy đủ tất cả các danh mục thông tin tiềm ẩn này là rất khó hoặc bất khả thi. Tuy nhiên, các thông tin mà tiếng nói chắc chắn chứa đựng bao gồm:

-   **Nội dung ngôn ngữ (văn bản)** của tín hiệu tiếng nói có thể chứa đựng bất cứ thông tin gì:
    -   Nếu bạn tiết lộ thông tin riêng tư về bản thân trong một cuộc trò chuyện, thì rõ ràng thông tin đó sẽ xuất hiện trong bản phiên âm của cuộc hội thoại.
    -   Cách lựa chọn từ ngữ và phong cách nói chuyện [có thể tiết lộ nhiều điều về người nói](https://www.theguardian.com/lifeandstyle/2009/mar/08/language-voice) mà chính bản thân họ không hề nhận ra.
-   **Nội dung cận ngôn ngữ** (para-linguistic, tức là phi ngôn ngữ/phi văn bản) chứa một lượng thông tin phong phú:
    -   Danh tính người nói (speaker identity)
    -   Các đặc điểm thể chất, bao gồm giới tính, kích thước cơ thể và độ tuổi
    -   Trạng thái cảm xúc (vui, buồn, tức giận, phấn khích, v.v.)
    -   Phong cách nói (trang trọng, thân mật, sân khấu, v.v.)
    -   Trạng thái sức khỏe, chẳng hạn như cảm cúm, sức khỏe tâm thần và các bệnh như [Alzheimer](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4876915/), cũng như sự mệt mỏi và tình trạng say xỉn.
    -   Sự liên kết và thuộc về các nhóm tham chiếu, bao gồm các nhóm bản dạng giới, sắc tộc, nền tảng văn hóa, vùng miền, quốc tịch, các mối liên kết chính trị và tôn giáo, v.v.
-   **Trong tương tác với những người nói khác:**
    -   Bằng chứng cho thấy bạn đã gặp gỡ và trò chuyện với người nói kia
    -   Mức độ quen thuộc, thân mật và [sự tin cậy](https://arxiv.org/abs/2007.15711).
    -   Cấu trúc quyền lực trong cuộc trò chuyện (người dẫn dắt/người nghe theo/đối tác)
    -   Các mối quan hệ gia đình, tình cảm và các mối quan hệ xã hội khác

Nói cách khác, tiếng nói chứa đựng hoặc có thể chứa đựng hầu như tất cả các loại thông tin riêng tư và nhạy cảm mà bạn có thể tưởng tượng ra. Vì tiếng nói là công cụ giao tiếp nên điều này không có gì ngạc nhiên; bất cứ điều gì chúng ta có thể truyền đạt đều có thể được nói ra. Ngược lại, nếu chúng ta nhận thấy rằng quyền riêng tư là quan trọng, thì tiếng nói là một trong những tín hiệu quan trọng nhất cần được bảo vệ.


## Các loại quyền riêng tư

Trong một phạm vi rộng hơn tiếng nói, quyền riêng tư có thể được phân thành bảy loại chính: {cite:p}`finn2013seven`

-   **Riêng tư của cá nhân (Privacy of the person)**, liên quan đến quyền riêng tư đối với cơ thể vật lý của chúng ta cũng như bất kỳ thông tin nào về cơ thể như dấu vân tay, đặc điểm giọng nói và lịch sử y tế.
-   **Riêng tư về hành vi và hành động (Privacy of behaviour and action)**, liên quan đến quyền riêng tư đối với những gì chúng ta làm; ví dụ, không ai cần biết tôi làm gì trong phạm vi ngôi nhà của mình.
-   **Riêng tư về giao tiếp (Privacy of communication)**, đặc biệt quan trọng trong giao tiếp tiếng nói và liên quan đến quyền riêng tư đối với nội dung và siêu dữ liệu (meta-content) của quá trình giao tiếp. Nghĩa là, nó không chỉ liên quan đến nội dung nghĩa đen của cuộc giao tiếp, mà còn về phong cách giao tiếp và thực tế là cuộc giao tiếp đó đã diễn ra.
-   **Riêng tư về dữ liệu, bao gồm hình ảnh và âm thanh (Privacy of data, including images and sound)**, mở rộng quyền riêng tư của tài sản đối với các thực thể phi vật chất. Đặc biệt, loại riêng tư này giải quyết vấn đề chia sẻ thông tin liên quan đến một bên thứ ba.
-   **Riêng tư về suy nghĩ và cảm xúc (Privacy of thoughts and feelings)**, nói cách khác là quyền riêng tư của việc tư duy. Chúng ta có quyền chia sẻ suy nghĩ và cảm xúc của mình với những người chúng ta muốn, hoặc chọn không chia sẻ cảm xúc với bất kỳ ai.
-   **Riêng tư về vị trí và không gian (Privacy of location and space)**, ngày càng trở nên quan trọng khi rất nhiều thiết bị di động của chúng ta có khả năng theo dõi vị trí của bạn. Tuy nhiên, loại riêng tư này bao gồm cả vị trí hiện tại và lịch sử vị trí (như trong việc theo dõi vị trí).
-   **Riêng tư về sự liên kết (Privacy of association)**, liên quan đến quyền riêng tư về việc bạn có thuộc về hoặc liên kết ở mức độ nào với một nhóm cụ thể (tôn giáo, chính trị, sắc tộc, bản dạng giới, nghề nghiệp, nhóm sở thích, v.v.).

Lưu ý rằng danh sách này không đưa ra bất kỳ tuyên bố nào liên quan đến *quyền* đối với các loại riêng tư này, mà chỉ chỉ ra rằng các vấn đề riêng tư thường có thể được chia nhỏ thành các chủ đề phụ này. Việc một người có quyền riêng tư hay không là quyết định ở cấp độ xã hội và lựa chọn chính trị, nơi các khía cạnh tâm lý và văn hóa đóng vai trò lớn.


## Các kịch bản đe dọa và tấn công

Các mối đe dọa đối với quyền riêng tư trong giao tiếp tiếng nói hầu như luôn có thể được định nghĩa là việc khai thác thông tin một cách bí mật cũng như lưu trữ, xử lý và sử dụng thông tin đó theo những cách mà người nói không biết và/hoặc không đồng ý. Sự đa dạng trong các kịch bản đe dọa chủ yếu phụ thuộc vào loại thông tin liên quan cũng như các bên liên quan. Cụ thể:

-   **Các công ty** có thể trích xuất thông tin từ người dùng cá nhân để trục lợi phi đạo đức:
    -   Các hợp đồng bảo hiểm và thế chấp có thể bị từ chối dựa trên thông tin được trích xuất một cách bí mật.
    -   Giá dịch vụ có thể bị tăng đối với các nhóm đối tượng dễ bị tổn thương.
    -   Quyền truy cập thông tin (kết quả tìm kiếm) có thể bị hạn chế và thông tin có thể bị nhắm mục tiêu (quảng cáo) để tác động bí mật đến người dùng nhằm đạt được lợi thế phi đạo đức. Hành vi như vậy thường được quảng cáo dưới danh nghĩa tùy chỉnh dịch vụ theo sở thích của người dùng, nhưng không cung cấp cho người dùng bất kỳ công cụ nào để lựa chọn cách tùy chỉnh dịch vụ đó.
-   **Các cơ quan nhà nước** có thể sử dụng các biện pháp giám sát và hạn chế truy cập đối với công dân của họ:
    -   Các chế độ độc tài có thể theo dõi và nghe lén phe đối lập chính trị, những người bất đồng chính kiến và các nhóm khác như tôn giáo, sắc tộc và xu hướng tình dục.
    -   Ngoài ra còn có các hoạt động sử dụng hợp pháp của cảnh sát để giám sát trong các cuộc điều tra hình sự.
-   **Các cơ quan nhà nước** có thể sử dụng các biện pháp giám sát đối với công dân nước ngoài:
    -   Các đặc vụ có thể sử dụng công nghệ tiếng nói để nghe lén (từ xa) và trích xuất thông tin.
-   **Tội phạm** có thể đánh cắp thông tin và sử dụng nó để trục lợi:
    -   Đánh cắp danh tính
    -   Giới săn tin (paparazzi) có thể đánh cắp thông tin riêng tư của những người nổi tiếng
    -   Thông tin riêng tư có khả năng bị sử dụng để tống tiền
    -   Nội dung nhạy cảm có thể bị bán làm phương tiện giải trí
-   **Các cá nhân** có thể bí mật sử dụng công nghệ tiếng nói để nghe lén và trích xuất thông tin của người khác:
    -   Ví dụ, lịch sử lệnh của loa thông minh có thể cung cấp quyền truy cập vào các lệnh trong quá khứ của tất cả người dùng, ngay cả khi các lệnh thoại đó được thực hiện ở chế độ riêng tư.

## Các hướng tiếp cận để bảo vệ quyền riêng tư và cải thiện khả năng sử dụng của công nghệ tiếng nói

### Các khái niệm thiết kế cơ bản

Ít nhất là dưới góc nhìn của châu Âu, các khái niệm thiết kế sau đây được coi là nền tảng của một thiết kế tốt cho quyền riêng tư. Trên thực tế, chúng được quy định bắt buộc bởi Quy định chung về bảo vệ dữ liệu (GDPR) của Liên minh Châu Âu.

-   **Riêng tư theo mặc định (Privacy by default);** các hệ thống phải luôn mặc định ở cấu hình ít xâm phạm quyền riêng tư nhất. Các dịch vụ bổ sung mang tính xâm phạm hơn có thể được người dùng chủ động lựa chọn (opt-in) nếu họ muốn.
-   **Riêng tư từ khâu thiết kế (Privacy by design);** quyền riêng tư không nên là một ý tưởng được thêm vào sau cùng, mà các hệ thống phải được thiết kế ngay từ đầu để đảm bảo tính riêng tư. Cấu trúc tổng thể của hệ thống phải được lựa chọn sao cho hỗ trợ tốt nhất cho quyền riêng tư.
-   **Tối thiểu hóa dữ liệu (Data minimization);** dữ liệu chỉ có thể được trích xuất từ người dùng ở mức độ tối thiểu cần thiết cho dịch vụ. Ví dụ, nếu bạn đặt một chiếc bánh pizza qua cổng thông tin điện tử, bạn cần cho biết loại bánh pizza nào bạn muốn và nơi nó sẽ được giao đến. Nếu không, bạn không thể nhận được dịch vụ. Nói cách khác, tất cả các dịch vụ đều yêu cầu một mức độ chuyển giao thông tin nhất định, nhưng các dịch vụ không thể yêu cầu các thông tin không liên quan. Ví dụ, dịch vụ giao pizza không thể hỏi bạn về bản dạng giới của bạn. Việc tối thiểu hóa dữ liệu cũng có thể được hiểu là *sinh thái học dữ liệu (data ecology)*, nhấn mạnh thực tế là "nhiều dữ liệu hơn" thường đồng nghĩa với việc tiêu thụ năng lượng và các tài nguyên khác nhiều hơn. Trên thực tế, dữ liệu có thể (hoặc nên) được coi chính là một tài nguyên thiên nhiên. Dòng lập luận này kết nối quyền riêng tư với [các mục tiêu phát triển bền vững của Liên Hợp Quốc](https://www.un.org/development/desa/disabilities/envision2030.html). Một khía cạnh quan trọng của tối thiểu hóa dữ liệu là thông tin chỉ nên được lưu trữ trong thời gian cần thiết cho dịch vụ.
-   **Sự đồng ý có thông tin và có ý nghĩa (Informed and meaningful consent);** khi người dùng lựa chọn một dịch vụ, họ phải nhận được thông tin về các tác động của nó đối với quyền riêng tư một cách dễ hiểu và dễ tiếp cận, và nhà cung cấp dịch vụ phải nhận được sự đồng ý của người dùng mà không có bất kỳ hình thức ép buộc nào.

Thông thường, việc các nhà cung cấp dịch vụ có quyền truy cập vào dữ liệu tổng hợp (chẳng hạn như giá trị trung bình của tập hợp) là hợp lý, nhưng họ không được phép truy cập vào thông tin của từng cá nhân. Ví dụ, trong một trường hợp giả định, nhà vận hành loa thông minh có thể nhận được thông tin rằng 55% người dùng là nam giới, nhưng không thể biết giới tính của bất kỳ người dùng cụ thể nào.

### Các định nghĩa pháp lý liên quan

Một vấn đề trung tâm về quyền riêng tư đối với tín hiệu tiếng nói là khái niệm "có thể nhận dạng duy nhất" (uniquely identifiable). Các khung pháp lý như GDPR quy định rằng thông tin riêng tư là những dữ liệu mà người dùng cá nhân "có thể nhận dạng duy nhất", nhưng không có định nghĩa chính xác về ý nghĩa thực sự của nó. Nếu đối tác của bạn nhận ra giọng "Xin chào" của bạn trên điện thoại, điều đó có nghĩa là đối với cô ấy, giọng "Xin chào" của bạn là có thể nhận dạng duy nhất. Tuy nhiên, nếu bạn đưa 10.000 mẫu tiếng nói cho đối tác của mình, một trong số đó là giọng "Xin chào" của bạn, thì có khả năng lớn là đối tác của bạn sẽ không tìm thấy giọng của bạn trong đống mẫu đó. Một câu hỏi chưa có lời giải đáp là: "Kích thước của nhóm mà người dùng có thể nhận dạng duy nhất là bao nhiêu?".

Một khía cạnh chi tiết hơn là về mức độ tin cậy. Hướng tiếp cận nhận dạng người nói (speaker recognition) là tìm người nói có khả năng cao nhất trong một nhóm tham chiếu có kích thước $N$, trong khi xác thực người nói (speaker verification) cố gắng xác định xem liệu chúng ta, trong một khoảng tin cậy nào đó, có thể chắc chắn bạn là người mà bạn tuyên bố hay không. Dưới góc độ kỹ thuật, điều này có nghĩa là chúng ta muốn tìm người nói có khả năng cao nhất, nhưng với một biên độ đủ lớn so với tất cả các người nói khác. Theo chiều ngược lại, chúng ta cũng có thể sử dụng một ngưỡng thấp hơn; chúng ta có thể nói rằng một mối tương quan có ý nghĩa thống kê đã làm lộ quyền riêng tư của người dùng. Ví dụ, nếu chúng ta phát hiện ra rằng người nói là bạn hoặc cha/mẹ của bạn, thì chúng ta đã có một mối tương quan thống kê đáng kể, nhưng bạn vẫn chưa bị nhận dạng duy nhất.

Một cân nhắc khác là dữ liệu kèm theo (adjoining data); Giả sử có một bản ghi âm của người nói A, và bạn tình cờ biết rất rõ người nói A. Khi đó, bạn sẽ dễ dàng nhận ra giọng nói của A trong bản ghi âm đó. Nghĩa là, bạn có rất nhiều kinh nghiệm (dữ liệu lưu trữ) về cách phát âm của A, do đó việc nhận dạng A đối với bạn là dễ dàng. Điều đó có nghĩa là A có thể nhận dạng duy nhất trong bản ghi âm đó không? Suy cho cùng, A sẽ không thể nhận dạng được nếu bạn không biết A (= nếu bạn không có dữ liệu lưu trữ từ trước về A).
Một biến thể nhỏ của trường hợp trên là bản ghi âm của người nói B, nơi B là một người nổi tiếng công chúng, do đó các mẫu giọng nói của ông ấy đã có sẵn trên mạng. Điều đó có làm cho bản ghi âm của B trở nên có thể nhận dạng duy nhất không? Hoặc nếu có bản ghi âm của một người hiện chưa nổi tiếng C, người sau này trở nên nổi tiếng. Điều đó có làm thay đổi trạng thái của bản ghi âm C thành có thể nhận dạng duy nhất không?

Ngày nay, câu hỏi này vẫn chưa có lời giải đáp và chúng ta chưa có sự diễn giải thống nhất về ý nghĩa thực sự của "có thể nhận dạng duy nhất". Mức độ tin cậy thống kê nào được giả định? Mức độ dữ liệu kèm theo nào được giả định (theo nghĩa của GDPR có lẽ là: bất kỳ và tất cả dữ liệu đang tồn tại)? Nó có thể thay đổi theo thời gian nếu thông tin mới được công khai không (nhiều khả năng là: có)? Nó có thể thay đổi theo thời gian nếu các công nghệ mới được phát triển không (nhiều khả năng là: có)?

Với *giả danh hóa* (pseudonymization), chúng tôi đề cập đến các phương pháp tương tự, nơi thông tin riêng tư được thay thế bằng một số thông tin khác. Ví dụ, chúng ta có thể thay thế thông tin nhận dạng cá nhân của người dùng *Alice* bằng một danh tính ảo *Adam*. Quá trình và phương pháp phân tách các luồng thông tin khác nhau này được gọi là *phân tách đặc trưng* (disentanglement).

### Riêng tư vi sai (Differential privacy)

Ngay cả khi hoạt động với dữ liệu tổng hợp, như độ tuổi trung bình của người dùng, việc trích xuất thông tin riêng tư trong một số tình huống vẫn khả thi. Ví dụ, nếu chúng ta biết độ tuổi trung bình của người dùng và số lượng người dùng tại thời điểm $t$, và chúng ta cũng biết rằng tuổi của người dùng $X$ được thêm vào dữ liệu tại thời điểm $t+1$, cùng với độ tuổi trung bình mới tại $t+1$, thì chúng ta có thể suy ra chính xác tuổi của người dùng $X$ bằng các phép toán đại số cơ bản. Như một biện pháp bảo vệ chống lại các cuộc tấn công vi sai như vậy, nhằm cung cấp [riêng tư vi sai](https://en.wikipedia.org/wiki/Differential_privacy) (differential privacy), hệ thống có thể thêm nhiễu vào bất kỳ hoạt động chuyển giao dữ liệu nào. Các điểm dữ liệu cá nhân sau đó bị làm mờ và không thể khôi phục chính xác. Tuy nhiên, giá trị trung bình của tập hợp vẫn có thể được suy ra nếu biết trước phân phối của nhiễu được thêm vào.

### Mã hóa đồng cấu (Homomorphic encryption)

Giả sử một nhà cung cấp dịch vụ có một mô hình độc quyền, ví dụ như phương pháp phân tích bệnh Alzheimer từ giọng nói, và bác sĩ của bạn muốn phân tích giọng nói của bạn bằng phương pháp đó. Giọng nói của bạn là riêng tư, vì vậy bạn không muốn gửi nó cho nhà cung cấp dịch vụ bên thứ ba, nhưng nhà cung cấp dịch vụ cũng không muốn gửi mô hình của họ cho bạn. [Mã hóa đồng cấu](https://en.wikipedia.org/wiki/Homomorphic_encryption) cung cấp một giải pháp để áp dụng mô hình bảo mật trên dữ liệu đã được *mã hóa*, sao cho bạn chỉ cần gửi dữ liệu của mình dưới dạng mã hóa cho nhà cung cấp dịch vụ. Bác sĩ của bạn sau đó sẽ chỉ nhận được chẩn đoán cuối cùng, còn nhà cung cấp không có được dữ liệu tiếng nói của bạn và bạn cũng không có được mô hình của họ. Về nguyên lý, khái niệm này rất đẹp đẽ, nó giải quyết bài toán không tin tưởng lẫn nhau một cách xuất sắc. Tuy nhiên, điểm hạn chế là các phương pháp mã hóa đồng cấu hiện tại yêu cầu tất cả các hàm xử lý phải được viết dưới dạng các hàm đa thức. Về mặt lý thuyết, chúng ta có thể chuyển đổi bất kỳ hàm nào thành đa thức tương ứng, nhưng sự gia tăng độ phức tạp tính toán thường là cực kỳ khủng khiếp. Do đó, các phương pháp bảo vệ quyền riêng tư dựa trên mã hóa đồng cấu thường có độ phức tạp tính toán quá cao để áp dụng thực tế. {cite:p}`armknecht2015guide`

Tuy nhiên, về nguyên tắc, người dùng có thể chọn lưu trữ dữ liệu riêng tư trên một thiết bị biên cục bộ, loại bỏ hoàn toàn nhà cung cấp dịch vụ lưu trữ khỏi vòng lặp.

Một rủi ro khác là trong mô hình myData, chúng ta thường giả định dữ liệu được lưu trữ tại một vị trí trung tâm duy nhất, nơi này sẽ trở thành một điểm yếu cốt tử (single point of failure). Nếu ai đó có được quyền truy cập bất hợp pháp vào kho lưu trữ này, toàn bộ dữ liệu của bạn sẽ bị xâm phạm. Do đó, việc phân tán dữ liệu đến nhiều địa điểm lưu trữ khác nhau có thể là giải pháp hợp lý hơn.


## Các mục tiêu thiết kế, giao diện người-máy và trải nghiệm người dùng

Một định kiến phổ biến là các yêu cầu về an ninh và quyền riêng tư gây ra khó khăn cho các nhà phát triển và làm cho các hệ thống trở nên phức tạp hơn đối với người dùng. Định kiến như vậy là không đúng và sai lầm một cách rõ ràng. Vấn đề là nhiều mối nguy hiểm về quyền riêng tư không hiển hiện đối với người quan sát thông thường và ảnh hưởng của chúng chỉ trở nên rõ ràng khi mọi việc đã quá muộn. Một lập luận khác là "*quyền riêng tư không phải là mối quan tâm của tôi vì tôi chưa thấy bất kỳ vấn đề nào về quyền riêng tư*", điều này tương tự như việc nói rằng "*tội phạm cưỡng bức không phải là mối quan tâm của tôi vì tôi chưa thấy vụ cưỡng bức nào*". Đây là một lập luận ngớ ngẩn. Các biện pháp bảo vệ quyền riêng tư được thiết kế để bảo vệ người dùng và nhà phát triển khỏi [những hậu quả rất tồi tệ](https://www.reuters.com/apple-cyber-idUSKCN1VR29K). Những vấn đề này là có thật. *Bạn* không thể phớt lờ chúng.

Tuy nhiên, thiết kế hướng tới quyền riêng tư *không chỉ* là bảo vệ người dùng. Nó cũng liên quan rất nhiều đến việc thiết kế công nghệ sao cho *dễ sử dụng* và trải nghiệm người dùng mang lại cảm giác trực quan và tự nhiên. Ví dụ, nhận dạng người nói có thể được sử dụng để cấp quyền truy cập vào công nghệ giọng nói sao cho người dùng không phải bận tâm với mật khẩu, mã PIN hoặc các phương pháp xác thực rườm rà khác. Nhìn chung, công nghệ tiếng nói hứa hẹn giúp truy cập các dịch vụ mà không cần phải cuộn qua các menu phức tạp trên máy giặt của bạn chỉ để tìm một chế độ được tối ưu hóa cho rèm cửa màu trắng làm bằng cotton.

Mục tiêu thiết kế tổng thể là làm sao để mọi người có thể *tin tưởng* hệ thống. Cụ thể, một hệ thống đáng tin cậy sẽ cần đạt được các yếu tố sau: {cite:p}`Chen2003,Xie2009`

-   **Nhất quán (consistent);** Thiết bị thực hiện đúng những gì nó tuyên bố, và tuyên bố đúng những gì nó thực hiện.
-   **Năng lực (competent);** Thiết bị có khả năng thực hiện những gì nó nên làm và những gì nó tuyên bố thực hiện.
-   **Thiện chí (benevolent);** Thiết bị quan tâm đến người dùng và thể hiện điều đó qua cả *những gì* nó nói và *cách* nó nói.

Các mục tiêu này được minh họa tốt nhất bằng các ví dụ sau:

-   Chúng ta nên tránh sự bất hòa nhận thức (cognitive dissonance); nếu máy tính có năng lực trí tuệ tương đương với một đứa trẻ 3 tuổi, nhưng lại nói bằng giọng nói uy quyền của một người trung niên, nó đang đưa ra các tín hiệu trái ngược có thể gây nhầm lẫn cho người dùng. Tương tự như vậy, nếu một máy tính làm rò rỉ toàn bộ thông tin của bạn ra thế giới trong khi trò chuyện bằng một giọng nói thân mật, nó cũng đang tạo ra ấn tượng sai lệch cho người dùng. Ngược lại, ấn tượng trực quan mà một thiết bị mang lại phải nhất quán với bản chất thực sự của nó.
-   Sự nhất quán là vô cùng quan trọng và một sai lầm có thể phải trả giá rất đắt; Nếu người bạn Greg của bạn một lần kể về sự cố sức khỏe thầm kín của bạn cho những người bạn khác nghe, điều đó sẽ phủ bóng đen nghi ngờ lên toàn bộ tương lai và cả 10 năm tình bạn đã qua. Liệu anh ấy trước đây đã từng kể bí mật của mình cho họ chưa? Anh ấy đã kể bí mật của mình cho ai khác nữa? Phải mất một thời gian rất dài để hàn gắn những vết nứt lòng tin như vậy.


## Các tình huống khó xử về mặt đạo đức (Ethical dilemmas)

Dưới đây là danh sách các câu hỏi giả định có thể (và thực tế đã) phát sinh trong việc thiết kế các hệ thống vận hành bằng tiếng nói:

-   Nếu một thiết bị nghe thấy tiếng kêu cứu, nó có nên tự động gọi cảnh sát không, ngay cả khi điều đó vi phạm quyền riêng tư và lòng tin của người dùng?
-   Nếu một thiết bị nghe thấy các dấu hiệu của bạo lực gia đình nhưng không có bằng chứng trực tiếp, nó có nên gọi cảnh sát không? Mức độ bằng chứng nào là bắt buộc?
-   Nếu một thiết bị hoặc dịch vụ nhận diện được bạn mắc bệnh Alzheimer hoặc một số bệnh nghiêm trọng khác, nó có nên nói cho bạn biết không? Ngay cả khi bác sĩ của bạn chưa nhận ra điều đó? Ngay cả khi chẩn đoán có thể bị sai? Ngay cả khi bạn chọn không muốn nghe điều đó? Ngay cả khi bạn có tiền sử trầm cảm và tin xấu đó có thể kích hoạt ý nghĩ tự tử?
-   Giả sử bạn đang mơ ước mua một chiếc xe đạp mới nhưng chưa kể với ai. Tuy nhiên, chiếc TV thông minh của bạn biết điều đó vì bạn đã tìm kiếm thông tin về chiếc xe đạp đó. Giả sử vợ/chồng của bạn đồng thời đang cố gắng chuẩn bị một món quà sinh nhật bất ngờ. Chiếc TV thông minh có nên gợi ý cho vợ/chồng bạn mua chiếc xe đạp đó không?
-   Giả sử quán cà phê địa phương của bạn có hệ thống gọi đồ uống tự động bằng giọng nói. Vào ngày này năm ngoái, bạn đã mua một món đồ uống bất ngờ mừng sinh nhật tại quán này. Máy tính có nên nhớ điều đó và chúc mừng sinh nhật bạn vào ngày hôm nay không?
-   Các dịch vụ càng hiểu rõ bạn thì chúng càng phục vụ bạn tốt hơn. Đó chắc chắn là một sự thật thực tế. (Việc các dịch vụ hiện tại chưa tối ưu không phải là một sự mâu thuẫn.) Tuy nhiên, bạn có muốn giữ quyền riêng tư trước các thiết bị giống như cách bạn giữ quyền riêng tư trước bạn bè của mình không? Ví dụ, bạn bè của bạn không đi theo bạn vào nhà vệ sinh hay phòng ngủ; liệu có ổn không nếu thiết bị của bạn làm điều đó?
-   Giả sử tất cả bạn bè của bạn đang sử dụng một dịch vụ "TakTik" để liên lạc trong nhóm. Giả sử thêm rằng dịch vụ TakTik này nổi tiếng với các hành vi xâm phạm quyền riêng tư. Khi lựa chọn dịch vụ, liệu bạn, với tư cách là một cá nhân, có thể đưa ra lựa chọn tự do dựa trên sở thích riêng tư của mình không? Nhìn chung, bạn có thể đưa ra những lựa chọn tự do về các dịch vụ truyền thông không?


## An ninh và quyền riêng tư trong nghiên cứu tiếng nói

Nghiên cứu khoa học dựa trên các lập luận được hỗ trợ bởi bằng chứng thực nghiệm, trong đó bằng chứng trong khoa học tiếng nói là các bản ghi âm tiếng nói. Do đó, việc truy cập dữ liệu tiếng nói là một phần bắt buộc của nghiên cứu khoa học tiếng nói. Để có được kết quả đáng tin cậy, các nhà nghiên cứu độc lập phải có khả năng xác minh kết quả của nhau, đồng nghĩa với việc họ phải có quyền truy cập vào cùng một nguồn dữ liệu hoặc các nguồn dữ liệu giống hệt nhau về mặt thực tế. Dữ liệu chia sẻ là tiêu chuẩn vàng cho [nghiên cứu có khả năng tái lặp](https://en.wikipedia.org/wiki/Reproducibility) (reproducible research). Tuy nhiên, việc chia sẻ dữ liệu tiếng nói có thể gây ra nhiều vấn đề liên quan đến quyền riêng tư của người nói.

Khái niệm "có thể nhận dạng duy nhất" ở đây là chìa khóa. Nếu một cá nhân *không* thể nhận dạng duy nhất trong một tập dữ liệu, thì bạn được phép chia sẻ dữ liệu đó. Ngược lại, nếu bạn loại bỏ tất cả dữ liệu nhận dạng, bạn có thể chia sẻ dữ liệu tương đối tự do. Tuy nhiên, dưới góc độ các cuộc thảo luận ở trên, cần làm rõ rằng không dễ để xác định những gì cấu thành dữ liệu nhận dạng cũng như những gì làm cho dữ liệu đó trở nên có thể nhận dạng "duy nhất".

Một cân nhắc quan trọng thứ hai là *sự đồng ý* (consent). Những người có giọng nói được ghi âm phải được phép tự do lựa chọn xem họ có muốn tham gia hay không và lựa chọn đó phải rõ ràng; bạn cần hỏi họ cụ thể xem họ có muốn tham gia vào cuộc ghi âm hay không. Nghiên cứu cần có khả năng chứng minh rằng sự đồng ý đã được cấp, và do đó sự đồng ý đó phải được ghi chép và lưu trữ cẩn thận. Sự đồng ý cũng phải được đưa ra một cách tự do sao cho không có hình phạt rõ ràng hay ẩn giấu nào khi từ chối. Hơn nữa, nếu bất kỳ dữ liệu nhận dạng duy nhất nào của người tham gia được lưu trữ, thì người tham gia phải được phép rút lại sự đồng ý sau đó. Tuy nhiên, có một số ngoại lệ quan trọng đối với quy tắc này; quyền rút lại sự đồng ý có thể bị từ chối, ví dụ, nếu việc đó làm hỏng tính toàn vẹn của tập dữ liệu, chẳng hạn như:

-   Nếu việc rút lui của một người tham gia có thể làm lệch lệch kết quả (bias), thì đó *có thể* là cơ sở để từ chối việc rút lui. Ví dụ, nếu một tập dữ liệu được xây dựng sao cho nó đại diện cho một phân đoạn cân bằng của dân số (tỷ lệ nam và nữ, độ tuổi, nền tảng văn hóa và giáo dục khác nhau được chọn để khớp với dân số chung), thì chúng ta không thể loại bỏ bất kỳ người tham gia nào mà không làm hỏng phân phối thống kê của tập dữ liệu. Hơn nữa, những người hiểu biết nhiều hơn về vấn đề riêng tư có thể có xu hướng rút lại sự đồng ý của họ nhiều hơn, khiến tập dân số mẫu bị sai lệch.
-   Nếu việc rút lui của một người tham gia có thể gây nguy hại đến khả năng tái lặp của kết quả nghiên cứu, thì đó *có thể* là cơ sở để từ chối việc rút lui. Ví dụ, nếu một thuật toán học máy được sử dụng trên một tập dữ liệu, thì chúng ta chỉ có thể tái tạo thuật toán đó nếu chúng ta có sẵn *chính xác* cùng một dữ liệu đó. Điều này đặc biệt có vấn đề nếu tập dữ liệu tương đối nhỏ, nơi những thay đổi nhỏ trong tập dữ liệu có thể dẫn đến những hậu quả lớn đối với đầu ra.

Để có cơ sở hợp lý cho việc từ chối quyền rút lại sự đồng ý, các tập dữ liệu có thể được thiết kế cân bằng hoặc có quy mô tương đối nhỏ. Việc thu thập các tập dữ liệu cân bằng là một thực hành tốt trong mọi trường hợp, do đó đây không phải là một hạn chế mà thực tế có thể nâng cao chất lượng nghiên cứu. Ngược lại, dữ liệu tốt là dữ liệu cân bằng và đó phải là mục tiêu của chúng ta; một hệ quả là chúng ta *có thể buộc phải* từ chối quyền rút lại sự đồng ý. Việc tránh thu thập các tập dữ liệu quá lớn cũng là điều tốt dưới góc độ *tối thiểu hóa dữ liệu* và sinh thái học dữ liệu.

Trong yêu cầu chấp thuận, người thu thập dữ liệu nên nêu rõ mục đích của tập dữ liệu (tức là *ràng buộc mục đích sử dụng* - purpose binding). Ví dụ, một tập dữ liệu có thể được thu thập để phát triển các phương pháp phát hiện từ kích hoạt (wake-word) và sự đồng ý được nhận cho mục đích đó. Khi đó, *việc sử dụng cùng một tập dữ liệu đó cho các thử nghiệm phát hiện người nói hoặc phân tích y tế giọng nói là không được phép*. Tuyệt đối không. Do đó, một thực hành tốt là yêu cầu sự đồng ý theo một phạm vi đủ rộng, để các nhà nghiên cứu có sự linh hoạt nhất định trong việc sử dụng dữ liệu. Tuy nhiên, sự chấp thuận chung cho tất cả các mục đích nghiên cứu không phải là thực hành tốt. Đặc biệt, chúng tôi khuyến nghị loại trừ việc xử lý các thông tin nhạy cảm như sức khỏe, sắc tộc, thông tin chính trị nếu đó không phải là mục đích rõ ràng của tập dữ liệu (so sánh với tối thiểu hóa dữ liệu).

Nếu một tập dữ liệu về bản chất có chứa dữ liệu nhận dạng duy nhất, thì các nhà nghiên cứu cần áp dụng các lớp bảo vệ mạnh mẽ hơn. Cụ thể, thông thường các nhà nghiên cứu phải theo dõi ai có quyền truy cập vào dữ liệu, để đảm bảo ràng buộc mục đích và cho phép rút lại sự đồng ý. Điều này cũng có thể yêu cầu bất kỳ nhà nghiên cứu nào tải dữ liệu xuống phải ký một hợp đồng thỏa thuận với nhà cung cấp dữ liệu, quy định rõ các điều khoản sử dụng. Một hợp đồng như vậy có thể được yêu cầu trong mọi trường hợp, không chỉ với dữ liệu nhận dạng duy nhất.

Các dữ liệu như thông tin y tế, dữ liệu về trẻ em hoặc các nhóm đối tượng dễ bị tổn thương khác, các mối liên kết chính trị, tôn giáo và bản dạng giới, v.v. là đặc biệt *nhạy cảm*. Nếu tập dữ liệu của bạn chứa *bất kỳ* thông tin nào như vậy, bạn phải áp dụng các biện pháp bảo vệ nghiêm ngặt hơn. Trước hết, quyền truy cập vào dữ liệu đó trên thực tế luôn phải được giới hạn cho những người có tên trong một hợp đồng ràng buộc pháp lý quy định rõ quyền truy cập và các mục đích sử dụng, xử lý và lưu trữ được phép.

Như một nguyên tắc chung, cần lưu ý rằng trưởng nhóm nghiên cứu (principal investigator) là người chịu trách nhiệm pháp lý đối với việc sử dụng dữ liệu được thu thập, lưu trữ và xử lý. Cụ thể, nếu một bên thứ ba tải dữ liệu xuống và lạm dụng nó, ví dụ như phân tích thông tin sức khỏe ngay cả khi không có sự đồng ý nào được thu thập cho mục đích đó, thì chính trưởng nhóm nghiên cứu là người chịu trách nhiệm. Tuy nhiên, trưởng nhóm nghiên cứu chỉ được yêu cầu áp dụng các *biện pháp bảo vệ hợp lý* để đảm bảo dữ liệu không bị lạm dụng. Mức độ bảo vệ thế nào là đủ hiện vẫn chưa có sự thống nhất chung. Nhiều khả năng là sẽ không bao giờ có các quy tắc quy định chính xác mức độ bảo vệ đủ.

Trong cuộc thảo luận ở trên, rõ ràng là bản chất của *khả năng nhận dạng duy nhất* có thể thay đổi theo thời gian khi có thông tin mới được công bố và các công nghệ mới xuất hiện. Điều này có nghĩa là các tập dữ liệu trước đây được bảo vệ đầy đủ có thể theo thời gian bị lộ các vấn đề riêng tư. Do đó, điều quan trọng là các nhà nghiên cứu phải liên tục giám sát các tập dữ liệu đã công bố của họ theo thời gian để nếu có các mối đe dọa mới xuất hiện, họ có thể có hành động xử lý thích hợp. Ví dụ, họ có thể rút lại hoàn toàn một tập dữ liệu. Các cách hợp lý để thực hiện điều này có thể bao gồm:

-   **Ngày hết hạn (Expiry date);** Tất cả các tập dữ liệu nên có một thời hạn sử dụng được nêu rõ ràng và việc sử dụng tập dữ liệu sau ngày hết hạn phải bị cấm. Người quản lý tập dữ liệu có thể cập nhật ngày hết hạn nếu không phát hiện mối đe dọa mới nào. Các công bố học thuật dựa trên các tập dữ liệu đã hết hạn không nên được chấp nhận.
-   **Kiểm soát quyền truy cập vào tập dữ liệu;** Để thực thi ràng buộc mục đích và cho phép rút lại tập dữ liệu khi cần thiết, người quản lý dữ liệu có thể yêu cầu tất cả người dùng phải đăng ký tài khoản và ký một hợp đồng chính thức quy định các mục đích sử dụng được chấp nhận.

Như một giải pháp cuối cùng, khi dữ liệu quá nhạy cảm và riêng tư đến mức không thể công bố rộng rãi, hệ thống có thể yêu cầu xử lý dữ liệu tại chỗ (on-site). Ví dụ, bạn có thể thiết kế một kiến trúc tính toán nơi dữ liệu nằm trên một máy chủ bảo mật, nơi các nhà nghiên cứu có quyền truy cập thông qua một giao diện lập trình ứng dụng ([API](https://en.wikipedia.org/wiki/Application_programming_interface)) an toàn. Dữ liệu không bao gồm đầu ra thô sẽ không bao giờ rời khỏi máy chủ, nhờ đó quyền riêng tư luôn được bảo toàn. Để có mức độ bảo mật cao hơn nữa, dữ liệu có thể được lưu trữ trên một hệ thống máy tính [cách ly vật lý](https://en.wikipedia.org/wiki/Air_gap_%28networking%29) (air-gapped computer system), có nghĩa là việc truy cập dữ liệu yêu cầu nhà nghiên cứu phải đến trực tiếp phòng đặt máy tính (không có kết nối mạng). Mức độ bảo mật này thường là đặc khu của các hệ thống cấp độ quân sự.
