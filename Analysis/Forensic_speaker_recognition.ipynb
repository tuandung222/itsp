(forensic-analysis)=
# Nhận dạng người nói trong khoa học pháp y (Forensic Speaker Recognition)

## Giới thiệu về nhận dạng người nói trong khoa học pháp y

Nhận dạng người nói trong khoa học pháp y (còn được gọi là so sánh giọng nói pháp y - Forensic voice comparison) nhằm mục đích áp dụng các phương pháp nhận dạng người nói để hỗ trợ thẩm phán hoặc bồi thẩm đoàn trong một phiên tòa xét xử. Theo nghĩa đó, từ *pháp y* (forensic) đề cập đến việc sử dụng các phương pháp khoa học hoặc công nghệ trong tố tụng pháp lý. Thật vậy, từ pháp y thường bị hiểu lầm là chỉ liên quan đến các vụ án bạo lực như giết người hay sát hại, có lẽ do ảnh hưởng từ phim ảnh và các chương trình truyền hình. Tuy nhiên, ý nghĩa của từ này rộng hơn nhiều.

Có hai ứng dụng chính của khoa học pháp y. Thứ nhất là khoa học pháp y *điều tra* (investigative forensic science), và thứ hai là khoa học pháp y *đánh giá* (evaluative forensic science). Khoa học pháp y điều tra nhằm mục đích sử dụng khoa học và công nghệ để hỗ trợ quá trình điều tra của cảnh sát, tức là, nói một cách đại thể, là quá trình từ khi khám nghiệm hiện trường vụ án cho đến khi bắt giữ nghi phạm, hoặc tìm ra một tang vật nghi vấn. Khoa học pháp y đánh giá sẽ tham gia vào quá trình tố tụng khi các chứng cứ thu thập được trong giai đoạn điều tra được sử dụng để hỗ trợ thẩm phán hoặc bồi thẩm đoàn trong quá trình ra phán quyết của một phiên tòa (chẳng hạn như trả lời câu hỏi "Nghi phạm vô tội hay có tội?").

Một ví dụ dưới đây sẽ giúp làm rõ các khái niệm này. Hãy tưởng tượng một vụ án trong đó một người (gọi là *thủ phạm*) lén lút lấy trộm một số đồ đạc từ một ngôi nhà, hành vi này được coi là tội phạm hình sự theo một hệ thống pháp luật cụ thể. Ban đầu, danh tính của thủ phạm không được biết và cảnh sát không có bất kỳ manh mối nào về việc ai có thể là tác giả của vụ trộm. Sau đó, chủ nhà gọi điện báo cảnh sát và làm đơn trình báo chính thức, bắt đầu quá trình điều tra vụ án của cảnh sát.

- Ngay từ những hành động đầu tiên của cảnh sát, quy trình pháp y đã bắt đầu ở giai đoạn *điều tra*. Trong giai đoạn này, cảnh sát hành động phối hợp chặt chẽ với hệ thống tư pháp tương ứng, thường là dưới sự chỉ đạo của một thẩm phán. Do đó, với sự cho phép của thẩm phán, cảnh sát vào nhà và khám nghiệm hiện trường vụ án để tìm kiếm manh mối (quy trình này được gọi là Điều tra hiện trường vụ án, viết tắt là CSI). Giả sử người đột nhập vào nhà đã vô tình để lại dấu vân tay tại một trong những nơi cất giữ đồ đạc bị mất cắp. Cảnh sát sau đó sử dụng các kỹ thuật thích hợp để thu thập dấu vết vân tay đó và lưu trữ lại (ví dụ, bằng một bức ảnh chụp có độ phân giải cao). Tiếp theo, dấu vân tay này được so sánh với dấu vân tay của những người sống trong nhà, được thu thập qua việc lăn mực trên giấy hoặc sử dụng thiết bị quét trực tiếp (livescan). Cảnh sát nhận thấy có những khác biệt cơ bản giữa dấu vân tay tại hiện trường và tất cả các dấu vân tay của những người sống trong nhà, dẫn đến kết luận chắc chắn rằng dấu vân tay đó là của thủ phạm. Dấu vân tay này sau đó được gửi đến phòng thí nghiệm pháp y để truy tìm nghi phạm. Tại phòng thí nghiệm pháp y (thường thuộc lực lượng cảnh sát, nhưng không phải luôn luôn), dấu vân tay được so sánh với một cơ sở dữ liệu vân tay bằng Hệ thống nhận dạng vân tay tự động (AFIS). Các hệ thống này tìm kiếm trong một cơ sở dữ liệu lớn các dấu vân tay đã đăng ký để tìm ra các mẫu có độ tương đồng cao. Hệ thống AFIS cho ra một danh sách các ứng viên, tức là các dấu vân tay rất gần với dấu vân tay tìm thấy tại hiện trường vụ án (còn gọi là *truy vấn* - query hoặc *mẫu thử* - probe). Nhờ danh sách xếp hạng này, cảnh sát có được một danh sách thu hẹp các đối tượng tình nghi để xem xét. Tiếp theo, cảnh sát đánh giá các thông tin khác liên quan đến vụ án (ngoại phạm, nhân chứng, tiền án tiền sự) đối với các ứng viên đó, và cuối cùng, cảnh sát bắt giữ một nghi phạm và người này được đăng ký trong hệ thống tư pháp. Giai đoạn điều tra kết thúc.

- Sau đó, quy trình pháp y tiếp tục chuyển sang giai đoạn *đánh giá*. Lúc này, thẩm phán đã có một nghi phạm, và do đó ấn định ngày bắt đầu phiên tòa xét xử, nơi sẽ quyết định xem nghi phạm có phải là tác giả của dấu vân tay tại hiện trường vụ án hay không. Để đưa ra quyết định này, thẩm phán có hai nguồn thông tin chính. Thứ nhất là *chứng cứ pháp y*, dưới dạng dấu vân tay tìm thấy tại hiện trường vụ án và mẫu vân tay đối chứng của nghi phạm. Thứ hai là tất cả các thông tin khác do cảnh sát báo cáo từ giai đoạn điều tra (tiền án tiền sự, nhân chứng, v.v.). Mục tiêu của giai đoạn đánh giá pháp y này là so sánh dấu vân tay tại hiện trường với dấu vân tay tương ứng của nghi phạm. Tuy nhiên, các thẩm phán không thể tự mình so sánh hai mảnh chứng cứ này vì họ không có chuyên môn kỹ thuật. Do đó, thẩm phán giao nhiệm vụ này cho một chuyên gia giám định pháp y (thường là một cảnh sát, nhân viên của phòng thí nghiệm pháp y công hoặc tư, hoặc một giám định viên độc lập, tùy thuộc vào quốc gia và hệ thống pháp luật). Từ quá trình so sánh này, thẩm phán nhận được một báo cáo giám định pháp y, giúp ông (hoặc bồi thẩm đoàn) đưa ra quyết định của mình. Tất nhiên, trong quá trình đưa ra phán quyết, thẩm phán có thể triệu tập bất kỳ bên nào liên quan để đưa ra lời khai (nhân chứng, nghi phạm và cả giám định viên pháp y) nhằm làm rõ các vấn đề trước khi ra quyết định.

Ví dụ về dấu vân tay này có thể dễ dàng ngoại suy sang các tín hiệu tiếng nói. Các trường hợp điển hình có thể thu thập được chứng cứ tiếng nói là các thiết bị ghi âm ẩn, các cuộc điện thoại bị nghe lén, các video được ghi lại hoặc các tin nhắn trong hộp thư thoại. Chúng ta gọi tài liệu tiếng nói này là tiếng nói *nghi vấn* (questioned) hoặc tiếng nói *chưa biết* (unknown). Nếu tiếng nói thu thập được trong bất kỳ tình huống nào chứa các manh mối tội phạm, một cuộc điều tra có thể bắt đầu và có thể kết thúc bằng việc một nghi phạm bị bắt giữ và xét xử. Sau đó, nghi phạm sẽ được yêu cầu ghi âm một số mẫu tiếng nói, mà chúng ta gọi là tiếng nói của *nghi phạm* (suspect) hoặc tiếng nói *đã biết* (known). Trong giai đoạn đánh giá, giám định viên sẽ thực hiện so sánh giữa mẫu tiếng nói nghi vấn và mẫu tiếng nói đã biết của nghi phạm, và gửi một báo cáo kết luận cho thẩm phán.

Thực tế là chứng cứ tiếng nói rất phổ biến trong các phiên tòa hình sự. Điều này đã thúc đẩy nhiều phòng thí nghiệm pháp y đầu tư vào nhân sự và cơ sở hạ tầng cần thiết để thực hiện khoa học tiếng nói pháp y. Tuy nhiên, các phương pháp được sử dụng để phân tích tiếng nói nhằm thu thập thông tin ở giai đoạn điều tra hoặc đánh giá là cực kỳ đa dạng. Một số ví dụ được đưa ra dưới đây:

- Trong giai đoạn đầu của khoa học tiếng nói pháp y, chúng ta có thể tìm thấy các phương pháp gọi là *dấu giọng* (voiceprint), nhằm so sánh các phổ đồ (spectrogram) hoặc thậm chí các dạng sóng (waveform) để tìm kiếm trực quan các đặc trưng *khớp* nhau giữa giọng nói của nghi phạm và giọng nói của thủ phạm. Mặc dù đây là phương pháp ưa thích thường được trình chiếu trong các bộ phim truyền hình (có lẽ vì nó giống với những gì được thực hiện với dấu vân tay), nó đã bị loại bỏ rộng rãi như một phương pháp khoa học hợp lệ. Lý do là các đặc trưng phân biệt người nói qua giọng nói không dễ dàng nhận biết được bằng mắt thường, và cũng có sự biến thiên tiếng nói rất lớn có thể đánh lừa mắt người mà không liên quan gì đến danh tính của người nói. Ví dụ, hai phổ đồ hoặc dạng sóng của hai đoạn tiếng nói có thể trông giống nhau do nhiều yếu tố không liên quan đến danh tính người nói, chẳng hạn như nhiễu nền tương đồng trong tệp ghi âm nghi vấn và đã biết. Ngoài ra, việc tìm kiếm một sự *khớp* nhau hoàn hảo giữa các phổ đồ hoặc dạng sóng là điều không tưởng chính vì tất cả các biến thiên này. Do đó, phương pháp này hoàn toàn không được khuyến nghị.

- Có các phương pháp gọi là phương pháp thính giác (auditory methods), trong đó giám định viên trực tiếp lắng nghe các giọng nói và đưa ra báo cáo dựa trên kinh nghiệm cá nhân. Các phương pháp này ngày càng ít được sử dụng vì chúng thiếu minh bạch, không thể kiểm chứng (ngoài việc tin tưởng vào giám định viên hay không), và dễ bị ảnh hưởng bởi tính chủ quan, định kiến và các lỗi diễn giải. Do đó, phương pháp này cũng hoàn toàn không được khuyến nghị.

- Ngoài ra còn có các phương pháp gọi là *ngữ âm - âm học dựa trên con người* (human-based phonetic-acoustic methods), trong đó giám định viên thường là một nhà ngôn ngữ học tính toán hoặc nhà ngữ âm học, và việc phân tích được thực hiện bằng các phương pháp kỹ thuật dựa trên các đặc trưng ngữ âm - âm học trích xuất từ tín hiệu tiếng nói. Các đặc trưng điển hình được trích xuất trong nhóm phương pháp này bao gồm tần số formant, băng thông và quỹ đạo formant, nhưng nhiều thông tin ngôn ngữ và ngữ âm khác cũng có thể được sử dụng. Giám định viên thường được hỗ trợ bởi các chương trình máy tính để trích xuất các đặc trưng này (như bộ theo dõi formant hoặc bộ lọc nhiễu), nhưng các đặc trưng đó thường được nhà ngữ âm học lựa chọn thủ công. Sau đó, việc so sánh được thực hiện theo một mô hình thống kê nào đó, đôi khi được tự động hóa bằng một chương trình máy tính khác; hoặc thậm chí đưa ra quyết định khớp/không khớp mang tính chủ quan (quyết định chủ quan này không được khuyến nghị vì những lý do được giải thích bên dưới trong chương này). Mặc dù có sự hỗ trợ tính toán mạnh mẽ, toàn bộ quy trình vẫn đòi hỏi một nhà ngữ âm học hoặc ngôn ngữ học chuyên gia, và nhiều quyết định thiết kế cũng như báo cáo vẫn mang tính phụ thuộc nhiều vào con người.

- Cuối cùng, có các phương pháp tự động và bán tự động (automatic and semi-automatic methods), dựa trên việc sử dụng một hệ thống nhận dạng người nói tự động để so sánh hai tài liệu tiếng nói bằng các phương tiện tự động. Một cách khác, hệ thống nhận dạng người nói bán tự động yêu cầu một giai đoạn trích xuất đặc trưng trước đó có thể có sự tham gia của con người. Tuy nhiên, các thuật toán để thực hiện việc so sánh được tự động hóa hoàn toàn. Giám định viên sau đó có thể đánh giá chứng cứ bằng một mô hình thống kê, hoặc thậm chí đưa ra quyết định khớp/không khớp mang tính chủ quan (quyết định chủ quan này không được khuyến nghị vì những lý do được giải thích bên dưới trong chương này).

Trong chương này, chúng ta sẽ chỉ xem xét các phương pháp tự động. Các phương pháp bán tự động là sự mở rộng dễ dàng của các phương pháp tự động, với điểm khác biệt là hệ thống được cung cấp âm thanh đã được xử lý trước bởi con người (thường là để thực hiện phát hiện hoạt động tiếng nói, loại bỏ tạp âm hoặc phân đoạn người nói mục tiêu). Tuy nhiên, cả hai loại hệ thống đều có cùng một kiến trúc cho các mục đích của chương này.

Chúng ta sẽ không xem xét các phương pháp ngữ âm - âm học dựa trên con người vì công nghệ liên quan đến các phương pháp đó bị giới hạn, ngoài các công cụ được sử dụng bởi các giám định viên để theo dõi formant hoặc giảm nhiễu. Tuy nhiên, các phương pháp đó được đề cập trong các chương khác.

## Thách thức trong nhận dạng người nói tự động pháp y

Trong khoa học pháp y, các điều kiện của tín hiệu tiếng nói thường rất bất lợi. Tài liệu tiếng nói nghi vấn thường có thời lượng ngắn, điều kiện âm học không được kiểm soát (ví dụ: tiếng vang, môi trường âm học, v.v.), độ nhiễu cao, sự đa dạng của các kênh truyền dẫn và các lược đồ mã hóa tiếng nói (các mạng di động và cố định khác nhau, VoIP, microphone, v.v.), điều kiện cảm ứng cảm xúc và nỗ lực phát âm thay đổi, v.v. Mặt khác, tiếng nói của nghi phạm cũng có thể biểu hiện sự biến thiên như vậy, vì điều kiện của các bản ghi âm đã biết có thể đa dạng từ các cuộc nghe lén điện thoại không mang tính buộc tội nhưng nghi phạm thừa nhận mình là tác giả, cho đến tiếng nói thu được bằng microphone chất lượng cao trong phòng yên tĩnh, v.v. Môi trường này tạo ra sự không khớp và biến thiên rất lớn, và thường là đầy thử thách, trong các điều kiện của tiếng nói nghi vấn và tiếng nói của nghi phạm. May mắn thay, các công nghệ hiện đại cho nhận dạng người nói có thể xử lý sự biến thiên và sự không khớp này, nhưng chỉ ở một mức độ nhất định. Có thể thấy rõ rằng các điều kiện bất lợi điển hình trong một kịch bản pháp y có thể làm suy giảm nghiêm trọng hiệu suất của các hệ thống. Điều này làm cho việc thực hiện một quá trình xác thực (validation) thích hợp cho hệ thống trong các điều kiện pháp y, hoặc gần giống với chúng, là điều thiết yếu trước khi áp dụng vào công việc thực tế.

Ngoài ra, trong khoa học pháp y, các cơ sở dữ liệu rất khó thu thập và sử dụng, ngay cả đối với mục đích nghiên cứu. Mặc dù dữ liệu có thể phong phú ở một số phòng thí nghiệm pháp y, nhưng có những hạn chế về quyền riêng tư và pháp lý gây khó khăn cho việc truy cập công khai và chia sẻ các cơ sở dữ liệu lớn trong điều kiện pháp y thực tế. Điều này ở một khía cạnh nào đó gây trở ngại cho sự tiến bộ của lĩnh vực này.

## Kiến trúc cơ bản của một hệ thống nhận dạng người nói tự động

Hệ thống nhận dạng người nói điển hình tính toán một *điểm số* (score) từ việc so sánh hai tài liệu tiếng nói. Một tài liệu sẽ là tiếng nói nghi vấn từ một kẻ phạm tội. Tài liệu kia sẽ là tiếng nói từ nghi phạm. Đối với các mục đích của chương này, kiến trúc được tóm tắt như sau:

![figure_basicArchitecture.png](attachments/figure_basicArchitecture.png)

Điểm số được tính toán bởi tất cả các kỹ thuật đã được giải thích ở các chương khác, cụ thể là Mô hình hỗn hợp Gauss (GMM), Phân tích định vị tuyến tính xác suất (PLDA), i-Vector, x-Vector, v.v. Tuy nhiên, đối với mục đích của chương này, chúng ta sẽ thảo luận về các ứng dụng pháp y của các hệ thống đó, chứ không đi sâu vào cơ chế hoạt động bên trong của chúng. Do đó, đối với mục đích của chúng ta, kiến trúc cơ bản được mô tả ở trên là quá đủ.

Theo quy ước, điểm số thường được diễn giải theo cách sau: điểm số cao hơn chỉ ra rằng hai giọng nói có khả năng cao hơn được tạo ra bởi cùng một người nói. Điểm số thấp hơn chỉ ra rằng hai giọng nói có khả năng cao hơn được tạo ra bởi những người nói khác nhau. Lưu ý rằng các từ *cao hơn* (higher) và *thấp hơn* (lower) được sử dụng thay vì *cao* (high) và *thấp* (low). Điều này là do điểm số không nhất thiết phải nằm trong một phạm vi xác định trước hoặc được chuẩn hóa. Thuộc tính duy nhất và quan trọng nhất của điểm số là cái gọi là *năng lực phân biệt* (discriminating power), tức là khả năng phân biệt giữa các cặp giọng nói đến từ cùng một người và từ những người khác nhau. Với mục đích này, phạm vi biến thiên của các điểm số của một hệ thống là hoàn toàn tùy ý, hoặc thậm chí không quan trọng. Nói cách khác, không quan trọng nếu một hệ thống đưa ra điểm số từ 0 đến 1, hoặc từ 5000 đến 1000000. Điều quan trọng là điểm số của cùng một người sẽ cao hơn điểm số của những người khác nhau, và ngược lại.

Tất nhiên, điểm số được tính toán bởi bất kỳ hệ thống nào cũng không hoàn hảo, và một số so sánh giữa các tài liệu tiếng nói đến từ cùng một người có thể cho điểm số thấp hơn so với một số so sánh giữa các tài liệu tiếng nói từ những người khác nhau (và ngược lại). Điều này là bình thường trong các hệ thống thực tế: mọi hệ thống đều mắc *sai số*. Tuy nhiên, điều này càng đúng hơn trong khoa học pháp y, khi xem xét các điều kiện thường rất bất lợi của tiếng nói. Vì vậy, mặc dù hành vi này cần phải tránh, sự hoàn hảo không tồn tại, và chúng ta phải chấp nhận một số *sai số* này trong hoạt động thực tế.

Cần nhấn mạnh rằng nhận dạng người nói pháp y khác biệt đáng kể so với xác thực người nói (speaker verification hoặc authentication). Trong xác thực người nói, như đã giải thích ở nơi khác, điểm số được so sánh với một ngưỡng để đưa ra quyết định giữa các giả thuyết cùng một người và khác người. Đây là một cách sử dụng nhận dạng người nói điển hình trong các ứng dụng bảo mật (xác thực hoặc kiểm soát truy cập, còn gọi là sinh trắc học giọng nói) hoặc trong các ứng dụng thương mại (đặc trưng hóa người nói cho mục đích tiếp thị). Trong các hệ thống xác thực người nói này, các lỗi điển hình là chấp nhận sai (false acceptance - còn gọi là dương tính giả) và từ chối sai (false rejection - còn gọi là âm tính giả). Ngưỡng thường được tối ưu hóa để đối phó với một hàm chi phí phát hiện, và các quyết định được đưa ra tương ứng.

Tuy nhiên, xác thực người nói không phù hợp cho các ứng dụng pháp y. Cách thức chính xác để sử dụng kiến trúc cơ bản này cho mục đích pháp y được trình bày dưới đây, áp dụng cho cả mục đích điều tra và mục đích đánh giá.

## Nhận dạng người nói pháp y trong các ứng dụng điều tra

Kiến trúc cơ bản của hệ thống nhận dạng người nói thường được sử dụng cho các ứng dụng điều tra nhằm thu được danh sách xếp hạng các người nói ứng viên từ một cơ sở dữ liệu lưu trữ các bản ghi âm đã biết. Mục tiêu chính của các hệ thống này là xác định nghi phạm trước khi bắt giữ họ, hoặc thu hẹp danh sách ứng viên cho một phiên tòa. Điều này bắt chước việc sử dụng các hệ thống sinh trắc học khác như các hệ thống AFIS được áp dụng rộng rãi cho dấu vân tay. Chế độ hoạt động này được gọi là *nhận dạng người nói* (speaker identification) trong tài liệu nghiên cứu nhận dạng người nói, và tương tự như dấu vân tay, chúng ta có thể gọi thiết lập này là Hệ thống nhận dạng người nói tự động (Automated Speaker Identification System - ASIS).

Cách tích hợp kiến trúc cơ bản của hệ thống nhận dạng người nói cho mục đích điều tra được thể hiện như sau:

![figure_investigative.png](attachments/figure_investigative.png)

Mặc dù không phổ biến như AFIS, các hệ thống gọi là ASIS này được nhiều lực lượng cảnh sát sử dụng để hỗ trợ điều tra trước khi các đánh giá pháp y được tiến hành. Chúng hữu ích để xếp hạng danh sách ứng viên cho các cuộc điều tra tiếp theo, tạo thêm tài liệu tiếng nói của người nói để so sánh tại phiên tòa, giảm thiểu danh sách nghi phạm tiềm năng, v.v. Tất nhiên, sau tất cả các cuộc điều tra, thẩm phán hoặc bồi thẩm đoàn mới là những người có thẩm quyền quyết định nghi phạm có tội hay vô tội. Nhưng các hệ thống này giúp tìm kiếm và theo dõi các nghi phạm.

Cần lưu ý rằng sự thiếu phổ biến của ASIS (tiếng nói) so với AFIS (vân tay) liên quan đến sự khác biệt về năng lực phân biệt của cả hai phương thức sinh trắc học này trong khoa học pháp y. AFIS hoạt động cực kỳ tốt vì năng lực phân biệt rất cao của việc so sánh một dấu vân tay hiện trường với một dấu vân tay đối chứng. Điều này đặc biệt đúng khi con người hỗ trợ các hệ thống bằng cách trích xuất thông tin liên quan trong dấu vân tay hiện trường, và do đó tất cả các hệ thống AFIS thường là bán tự động. Tài liệu tiếng nói trong các cuộc điều tra cũng thường được tiền xử lý thủ công bởi con người trước khi đưa vào ASIS, nhưng năng lực phân biệt của chứng cứ tiếng nói thấp hơn nhiều so với dấu vân tay trong các kịch bản pháp y điển hình. Điều này làm giảm sự quan tâm đối với ASIS so với AFIS.

## Nhận dạng người nói pháp y trong các ứng dụng đánh giá

Việc đánh giá mối quan hệ giữa hai mảnh chứng cứ tại các phiên tòa xét xử đã là chủ đề thảo luận trong những năm qua. Từ cuộc thảo luận này, một số khuyến nghị từ Mạng lưới các Viện Khoa học Pháp y Châu Âu (ENFSI) đã được đưa ra. Tổ chức này tích hợp nhiều phòng thí nghiệm pháp y và lực lượng cảnh sát Châu Âu (không chỉ từ Liên minh Châu Âu mà còn từ Thụy Sĩ, Na Uy, v.v.). Do đó, không khó để đoán rằng ảnh hưởng của họ là rất đáng kể. Các khuyến nghị của ENFSI chỉ ra rằng cách thức logic để thực hiện đánh giá hai tài liệu chứng cứ, bao gồm cả tiếng nói, là thông qua một khung tỷ số khả tích (likelihood-ratio framework, hoặc khung LR). Tiếp theo, chúng tôi sẽ mô tả khung này ở đây, cần lưu ý rằng các cách tiếp cận khác (không được ENFSI khuyến nghị) cũng có thể được sử dụng để tạo ra các báo cáo đánh giá trong khoa học pháp y.

### Khung tỷ số khả tích (Likelihood Ratio Framework)

Từ góc độ logic hình thức, kết quả đầu ra của một phép so sánh giữa hai tài liệu tiếng nói nhằm mục đích đánh giá chứng cứ phải là một giá trị hỗ trợ cho một trong các giả thuyết liên quan đến vụ án pháp y. Khung tỷ số khả tích ủng hộ một khung quyết định logic, được thiết lập trên lý thuyết quyết định Bayes nổi tiếng. Hãy xem xét một vụ án pháp y, trong đó chứng cứ pháp y chứa tài liệu tiếng nói nghi vấn và tài liệu tiếng nói của nghi phạm. Chúng ta sẽ sử dụng *biến số* $E$ để biểu thị thông tin trích xuất từ chính chứng cứ đó, có giá trị được quan sát và đã biết. Trong trường hợp nhận dạng người nói bán tự động hoặc tự động, thông tin đó về chứng cứ được thu thập bằng cách sử dụng một hệ thống nhận dạng người nói, hoặc theo cách hoàn toàn tự động (không có sự can thiệp của con người), hoặc phổ biến nhất là theo cách bán tự động, điển hình là qua giai đoạn tiền xử lý tiếng nói có sự hỗ trợ của con người. Do đó, thông tin mà chúng ta có thể thu được từ chứng cứ chỉ là điểm số (score) thu được từ kiến trúc nhận dạng người nói điển hình. Và đó là lý do tại sao chúng ta biểu thị điểm số này là $E$.

Ngoài ra, gọi $H_1$ và $H_2$ là các giả thuyết liên quan, còn gọi là các *mệnh đề* (propositions), trong vụ án pháp y. Khi đó, thẩm phán hoặc bồi thẩm đoàn phải đưa ra quyết định nghiêng về một trong hai giả thuyết này. Có thể có nhiều hơn hai mệnh đề trong một vụ án, nhưng chúng ta sẽ chỉ giả định hai mệnh đề trong chương này cho đơn giản. Các giả thuyết, hoặc mệnh đề, có liên quan đến quyết định trong một vụ án pháp y có thể trông như thế này, ví dụ:

* $H_1$: *Tiếng nói nghi vấn và tiếng nói của nghi phạm bắt nguồn từ cùng một người (hoặc cùng nguồn)*
* $H_2$: *Tín hiệu tiếng nói hiện trường và mẫu tiếng nói đối chứng bắt nguồn từ những người khác nhau (hoặc các nguồn khác nhau)*

Các mệnh đề này như được phát biểu là khá đơn giản, có thể là đơn giản nhất, và được gọi là các mệnh đề *cấp nguồn* (source-level propositions). Có nhiều mệnh đề khác khả thi trong một vụ án pháp y, chẳng hạn như *Tiếng nói nghi vấn được nói qua điện thoại bởi nghi phạm* (liên quan đến hành vi nói qua điện thoại), hoặc *Tiếng nói nghi vấn bắt nguồn từ người anh em sinh đôi của nghi phạm*. Những mệnh đề này có thể có nhiều hệ quả hơn từ góc độ pháp lý, hoặc thậm chí làm phức tạp thêm việc tính toán của các mô hình đánh giá chứng cứ một cách đáng kể. Nhưng để minh họa, trong chương này định nghĩa cấp nguồn của chúng ta về $H_1$ và $H_2$ là đủ. Chúng ta sẽ nói rằng *biến số* giả thuyết $H$ có hai giá trị $\{H_1, H_2\}$. Biến số giả thuyết $H$ này là biến số không quan sát được của chúng ta, còn gọi là biến *ẩn* (latent variable). Điều này có nghĩa là chúng ta muốn biết giá trị của $H$, nhưng không thể. Do đó, chúng ta cần *suy luận* giá trị của nó, tức là chúng ta muốn thu thập tối đa thông tin có thể về $H$ và sử dụng nó theo cách cho phép chúng ta biết thêm về nó.

Cuối cùng, có thêm một biến số cần định nghĩa, thực tế là biến số trừu tượng nhất. Gọi $I$ là thông tin *nền* (background information) có sẵn trong vụ án không liên quan đến chứng cứ $E$. Thông tin nền này có thể bao gồm bất kỳ điều gì không liên quan đến chứng cứ nhưng có liên quan đến vụ án. Nó có thể bao gồm lời khai của nhân chứng, tiền án tiền sự của nghi phạm, hoàn cảnh của vụ án (ví dụ, những mâu thuẫn trước đó giữa nghi phạm và các nạn nhân liên quan đến tội phạm, v.v.), vân vân.

Trong lý thuyết quyết định Bayes, các quyết định được đưa ra dựa trên xác suất của các giả thuyết khác nhau dưới ánh sáng của tất cả thông tin trong vụ án. Nói cách khác, chúng ta muốn có xác suất của các giá trị khác nhau của biến số chưa biết cần quan tâm (tức là biến mệnh đề $H$), cho trước tất cả thông tin quan sát được (tức là thông tin về chứng cứ, $E$; và thông tin nền, $I$). Chúng ta sẽ sử dụng ký hiệu $P \left( H = H_1 \left. \right| E,I \right)$ và $P \left( H = H_2 \left. \right| E,I \right)$ để đại diện cho các xác suất đó. $H_1$ và $H_2$, trong dạng đơn giản nhất của chúng, là loại trừ lẫn nhau và vét cạn, và do đó $P \left( H_1 \left. \right| E,I \right) = 1 - P \left( H_2 \left. \right| E,I \right)$.

Nhưng lúc này, một vấn đề nảy sinh. Giá trị của $P \left( H_1 \left. \right| E,I \right)$ (và $P \left( H_2 \left. \right| E,I \right)$ không thể được tính toán trực tiếp bởi thẩm phán hoặc bồi thẩm đoàn, bởi vì thẩm phán và bồi thẩm đoàn không thể tự đánh giá bất kỳ chứng cứ tiếng nói nào bằng cách sử dụng hệ thống nhận dạng người nói tự động. Điều này không chỉ nằm ngoài chuyên môn hay năng lực của họ trong tác vụ đó, mà còn nằm ngoài thẩm quyền của họ trong quá trình tố tụng tư pháp. Mặt khác, chính thẩm phán hoặc bồi thẩm đoàn mới là những người phải đưa ra quyết định, chứ không phải ai khác. Nếu giá trị của $P \left( H_1 \left. \right| E,I \right)$ không thể được phân tách theo một cách nào đó, thì không có cách nào để tách biệt việc đánh giá chứng cứ khỏi phần thông tin nền còn lại của vụ án. Vậy làm thế nào để giải quyết vấn đề này?

May mắn thay, định lý Bayes nổi tiếng có thể được sử dụng để phân tách các xác suất này, dẫn đến biểu thức sau:

$$\frac{P \left( H_1 \left. \right| E,I \right)} {P \left(H_2  \left. \right| E,I\right)}= \frac{{P \left( {\left. E \right|H_1,I} \right)}}{{P\left( {\left. E \right|H_2,I} \right)}} \cdot \frac{P \left(H_1 \left. \right| I \right)} {P \left( H_2 \left. \right|I \right)} $$

Từ phương trình này, chúng ta có thể phân biệt các đại lượng sau:

* Xác suất tiên nghiệm (prior probabilities) $P \left( H_1 \left. \right| I \right)$ và $P \left( H_2 \left. \right| I \right)$, thuộc thẩm quyền của thẩm phán hoặc bồi thẩm đoàn, và phải được họ ấn định mà không xem xét thông tin về chứng cứ $E$. Tỷ số của hai xác suất bổ sung này được gọi là *tỷ số tiên nghiệm* (prior odds), đại diện cho tỷ số ủng hộ hoặc chống lại giả thuyết $H_1$. Như vậy, tỷ số có giá trị $x$ có thể được diễn giải là *giả thuyết $H_1$ có khả năng đúng cao hơn $x$ lần so với giả thuyết $H_2$*, hoặc *tỷ số là $x$ so với 1 ủng hộ $H_1$*. Tỷ số này cũng được sử dụng trong cá cược, và chúng trở nên phổ biến chính vì thực tế này.

* Xác suất hậu nghiệm (posterior probabilities) $P \left( H_1 \left. \right| E,I \right)$ và $P \left( H_2 \left. \right| E,I \right)$, cũng thuộc thẩm quyền của thẩm phán hoặc bồi thẩm đoàn (vì họ xem xét thông tin nền là thông tin quan sát được), và phải được tuyên bố có xem xét đến thông tin về chứng cứ $E$ và thông tin nền $I$. Tỷ số của hai xác suất bổ sung này được gọi là *tỷ số hậu nghiệm* (posterior odds), và được diễn giải theo cách tương tự như tỷ số tiên nghiệm, nhưng dưới ánh sáng của cả $E$ và $I$, thay vì chỉ mỗi $I$.

* Cái gọi là tỷ số khả tích (likelihood ratio - LR), là tỷ số của hai khả tích (likelihoods): $P \left( {\left. E \right|H_1,I} \right)$ và $P\left( {\left. E \right|H_2,I} \right)$. Lưu ý rằng LR này là đại lượng *cập nhật* tỷ số tiên nghiệm bằng thông tin về chứng cứ, cụ thể là $E$. Nói cách khác, nếu thẩm phán hoặc bồi thẩm đoàn gán một tỷ số tiên nghiệm nào đó cho vụ án (tức là trước khi chứng cứ được đánh giá bằng LR), và LR nhân với tỷ số đó, họ sẽ thu được tỷ số hậu nghiệm và có thể đưa ra quyết định. LR thuộc thẩm quyền của chuyên gia giám định pháp y, và thẩm phán hay bồi thẩm đoàn không phải tính toán nó. Biểu thức của LR do đó là:

$$ LR= \frac{{P \left( {\left. E \right|H_1,I} \right)}}{{P\left( {\left. E \right|H_2,I} \right)}} $$

Lưu ý rằng LR cũng được điều kiện hóa theo thông tin nền $I$. Nhưng trong trường hợp này, nó được xem xét để thiết lập các điều kiện mà $E$ phải có. Ví dụ, nếu thông tin nền $I$ bao gồm thực tế là người nói nghi phạm là một người hút thuốc, và giám định viên nghĩ rằng điều này có thể ảnh hưởng đến phương pháp của họ, thì chứng cứ phải được đánh giá có xem xét đến tiếng nói từ những người hút thuốc trong dữ liệu được sử dụng để huấn luyện mô hình. Chúng ta sẽ thảo luận về cách huấn luyện mô hình sau, nhưng tại thời điểm này, chỉ cần ghi nhớ rằng không có bất kỳ sự không nhất quán nào về mặt pháp lý hay thủ tục khi điều kiện hóa các khả tích của LR theo $I$.

Như vậy, trong khung tỷ số khả tích (LR), giá trị của chứng cứ (ví dụ: tiếng nói nghi vấn và tiếng nói của nghi phạm) chính là bản thân giá trị LR, và giá trị này phải được tính toán bởi chuyên gia giám định pháp y. Quy trình suy luận được phác thảo như sau:

Khung diễn giải dựa trên LR này mang lại nhiều ưu điểm:

* Nó cho phép các nhà khoa học pháp y đánh giá và báo cáo một giá trị có ý nghĩa cho trọng số của chứng cứ trước tòa, với một cách diễn giải phổ quát. Theo nghĩa đó, ví dụ $LR=5$ có nghĩa là *chứng cứ ủng hộ giả thuyết $H_1$ gấp $5$ lần so với giả thuyết $H_2$*. Nói cách khác, nó sẽ nhân tỷ số tiên nghiệm lên 5 lần. Điều này tương đương với việc nói rằng xác suất của $H_1$ cho trước $E$ và $I$ được tăng lên. Giá trị LR càng cao thì sự ủng hộ dành cho $H_1$ càng mạnh, và xác suất hậu nghiệm $P \left( H_1 \left. \right| E,I \right)$ sẽ càng cao đối với một giá trị cho trước của xác suất tiên nghiệm $P \left( H_1 \left. \right| I \right)$. Điều tương tự cũng áp dụng cho giả thuyết $H_2$ khi LR nhỏ hơn 1: nó sẽ ủng hộ $H_2$, làm giảm xác suất hậu nghiệm của $H_1$.

* Vai trò của giám định viên và thẩm phán/bồi thẩm đoàn được phân định rõ ràng, từ đó phân tách thông tin trong vụ án thành hai nguồn có thể được xử lý bởi nhân sự có thẩm quyền tương ứng: giám định viên đánh giá chứng cứ (tức là tính toán LR) và thẩm phán hoặc bồi thẩm đoàn gán tỷ số tiên nghiệm và đưa ra quyết định.

* Dưới khung này, các xác suất có thể được diễn giải như mức độ tin tưởng (degrees of belief), cho phép kết hợp các ý kiến chủ quan dưới dạng các xác suất trong quy trình suy luận một cách rõ ràng và khoa học. Nói cách khác, thẩm phán hoặc bồi thẩm đoàn có thể đưa ra tỷ số tiên nghiệm của họ theo cách hoàn toàn chủ quan, mà không cần phải tính toán xác suất một cách rõ ràng. Thực tế đây là một quy trình mà con người chúng ta vốn dĩ vẫn thực hiện trong cuộc sống hàng ngày (hãy nghĩ về việc gán tỷ số tiên nghiệm cho việc trời mưa hay không mưa vào lần tới khi bạn rời khỏi nhà, để quyết định xem bạn có mang theo ô hay không).

* Bản thân giá trị LR có thể tự diễn giải được, ngay cả khi tỷ số tiên nghiệm không được thiết lập bởi thẩm phán hoặc bồi thẩm đoàn. Nó được diễn giải như là sự *ủng hộ* (support) đối với một ý kiến đã được phát biểu trước đó (tức là tỷ số tiên nghiệm), có được từ việc phân tích thông tin trong chứng cứ, $E$. Nói cách cách khác, nếu LR$>1$ chứng cứ sẽ ủng hộ rằng $H=H_1$, và nếu LR$<1$ chứng cứ sẽ ủng hộ rằng $H=H_2$. Cần lưu ý rằng LR *ủng hộ* một ý kiến về $H$, nhưng bản thân LR *không phải* là một ý kiến về $H$. Ý kiến về $H$ được đại diện bằng xác suất của giả thuyết, hoặc tỷ số (odds). Do đó, không thể đưa ra phán quyết về giá trị của $H$ chỉ dựa trên giá trị của LR, vì các quyết định sẽ được đưa ra từ xác suất/tỷ số hậu nghiệm, chứ không phải từ các mức độ ủng hộ đơn thuần.

* Nếu quy trình tính toán LR dựa trên dữ liệu và một mô hình thống kê xác định, nó sẽ minh bạch, có thể lặp lại và có thể kiểm chứng được. Ngoài ra, nếu quy trình dựa trên các hệ thống tự động, rất nhiều phép đánh giá chứng cứ có thể được thực hiện tự động, cho phép xác thực thực nghiệm và đo lường hiệu suất dễ dàng hơn. Chúng ta sẽ đề cập đến điều này sau.

### Sử dụng các hệ thống nhận dạng người nói trong khung LR

Như đã đề cập trước đó, nếu chúng ta sử dụng một hệ thống nhận dạng người nói để đánh giá chứng cứ tiếng nói trong một vụ án, tất cả thông tin mà hệ thống đưa ra về mức độ ủng hộ cho các giả thuyết được cô đọng thành cái gọi là *điểm số* (score) từ kiến trúc nhận dạng người nói tiêu chuẩn của chúng ta. Do đó, việc diễn giải chứng cứ tiếng nói bằng cách sử dụng các hệ thống nhận dạng người nói yêu cầu trước hết điểm số phải được tính toán bởi hệ thống, cho ra giá trị cụ thể $E=e$. Sau đó, điểm số này sẽ được diễn giải bằng một tỷ số khả tích, hay nói cách khác, điểm số được biến đổi thành một giá trị LR bằng việc sử dụng một mô hình thống kê hoặc học máy nào đó. Quy trình này còn được gọi là **hiệu chuẩn điểm số (score calibration)**. Quy trình này được minh họa như sau:

![figure_calibration.png](attachments/figure_calibration.png)

Do đó, mục tiêu của các phương pháp LR với các hệ thống nhận dạng người nói là sử dụng một mô hình để biến đổi các điểm số thành các giá trị LR trong một vụ án pháp y. Trong phần này, một số thuật toán phổ biến nhất để tính toán LR từ các điểm số sinh trắc học sẽ được mô tả.

Một số vấn đề kỹ thuật cần được lưu ý, chẳng hạn như thực tế là hầu hết các điểm số nhận dạng người nói nằm trong một phạm vi liên tục, dẫn đến tỷ số của các khả tích được tính toán từ các hàm mật độ xác suất (probability density functions) chứ không phải các xác suất rời rạc. Tuy nhiên, giá trị *LR liên tục* này sẽ hoàn toàn phù hợp với khung quyết định Bayes bằng cách xem xét vấn đề kỹ thuật này.

Trước khi mô tả từng phương pháp chuyển đổi điểm số sang LR (còn gọi là các phương pháp hiệu chuẩn), cần phải nói rằng tất cả các phương pháp đó là các mô hình thống kê hoặc thuật toán học máy được huấn luyện bằng một số dữ liệu *huấn luyện*, còn được biết đến trong khoa học pháp y là **dữ liệu nền (background data)**. Dữ liệu này sẽ được sử dụng để tính toán các *điểm số huấn luyện*, các điểm số này sau đó được sử dụng để biến đổi điểm số trong vụ án pháp y thành giá trị LR. Dữ liệu nền này là thiết yếu trong khoa học pháp y, và đặc biệt là trong nhận dạng người nói pháp y. Nếu dữ liệu nền không được lựa chọn cẩn thận, LR thu được có thể là một LR *tồi*, theo nghĩa là nó sẽ dẫn đến các tỷ số hậu nghiệm *tồi*. Chúng ta sẽ thảo luận về vấn đề đó sau, nhưng tại thời điểm này, điều quan trọng cần ghi nhớ là việc lựa chọn dữ liệu nền là một vấn đề then chốt trong nhận dạng người nói pháp y.

### Tính toán LR sinh (Generative) từ các điểm số

Việc tính toán $LR$ trong nhận dạng người nói pháp y theo truyền thống được thực hiện bằng cách sử dụng các kỹ thuật sinh (generative) mô hình hóa riêng biệt tử số và mẫu số của LR, cụ thể là $P \left(\left. E \right|H_1, I\right)$ và $P \left(\left. E \right|H_2, I\right)$. Điều này được thực hiện bằng cách trước tiên gán các mật độ xác suất cho chứng cứ bằng cách sử dụng các điểm số huấn luyện của cùng một người để thu được $P \left(\left. E \right|H_1, I\right)$, và gán các mật độ xác suất cho chứng cứ bằng cách sử dụng các điểm số huấn luyện của những người khác nhau để thu được $P \left(\left. E \right|H_2, I\right)$. Sau đó, giá trị của $E$ được cụ thể hóa bằng điểm số thu được trong vụ án pháp y (tính từ chứng cứ tiếng nói), và LR sẽ là tỷ số của hai mật độ xác suất cụ thể hóa tại giá trị đó.

Việc thiết kế phép biến đổi điểm số sang LR này có thể liên quan đến việc lựa chọn họ mật độ xác suất được sử dụng. Lựa chọn đơn giản nhất cho các điểm số nhận dạng người nói là phân phối Gauss, thu được thông qua Ước lượng khả tích cực đại (Maximum Likelihood - ML) từ tập hợp các điểm số huấn luyện. Tuy nhiên, điều này đòi hỏi các điểm số huấn luyện liên quan phải nhất quán với các hàm mật độ xác suất Gauss, điều này thường không phải là trường hợp thực tế. May mắn thay, một số phương pháp chuẩn hóa điểm số thường được sử dụng trong nhận dạng người nói giúp Gauss hóa các điểm số, và do đó các mô hình Gauss thường hoạt động khá tốt khi các phép chuẩn hóa điểm số này được thực hiện trong hệ thống nhận dạng người nói. Các phương pháp tiếp cận khác cho việc khớp mô hình sinh ML bao gồm việc sử dụng các Hàm mật độ nhân (Kernel Density Functions - còn gọi là cửa sổ Parzen), Mô hình hỗn hợp Gauss (GMM) và các phân phối tham số khác. Các mô hình Bayes cũng đã được áp dụng thành công khi điểm số huấn luyện khan hiếm.

### Tính toán LR sử dụng hồi quy logistic (Logistic Regression)

Hồi quy logistic là một kỹ thuật phân biệt phân loại (discriminative classification) nổi tiếng của học máy, được sử dụng rộng rãi cho nhiều bài toán phân loại. Trong bối cảnh nhận dạng người nói pháp y, mục tiêu của hồi quy logistic là thu được logarit của LR, tức là $\log(LR)$, dưới dạng một phép biến đổi affine (tức là dịch chuyển và tỷ lệ) của điểm số. Để đạt được điều này, hồi quy logistic không quan tâm đến tử số hay mẫu số của LR một cách riêng biệt. Thay vào đó, nó tối ưu hóa một hàm mục tiêu nhằm chỉ ra xem $\log(LR)$ thu được là tốt hay không. Hàm mục tiêu này được gọi là **hàm entropy chéo (cross-entropy function)**, được tính toán bằng cách sử dụng các điểm số huấn luyện, và nó được biết đến rộng rãi trong học máy (ví dụ, các mạng nơ-ron liên kết đầy đủ cho phân loại thường được huấn luyện với cùng mục tiêu entropy chéo này). Do đó, entropy chéo trong cơ sở dữ liệu huấn luyện được tối ưu hóa theo các kỹ thuật hạ cực gradient (gradient descent), và mô hình được khớp.

Hồi quy logistic được sử dụng rộng rãi trong nhận dạng người nói pháp y vì các thuộc tính khái quát hóa tốt, tính đơn giản và hiệu suất cao của nó.

### Phép biến đổi điểm số sang LR bằng thuật toán PAV (Pool Adjacent Violators)

Một cách tiếp cận khác để tính toán LR từ điểm số đã được đề xuất bằng cách sử dụng thuật toán PAV (Pool Adjacent Violators), thuật toán này có thể được huấn luyện để biến đổi một tập hợp các điểm số thành một tập hợp các giá trị LR thể hiện hiệu suất tối ưu trên tập huấn luyện. Do đó, một phép biến đổi PAV có thể được huấn luyện trên tập hợp các điểm số huấn luyện, và sau đó áp dụng cho một điểm số trong vụ án pháp y. Cũng có các biến thể của PAV, được lấy cảm hứng từ các phương pháp Bayes, giới hạn độ mạnh của LR thu được (tức là giới hạn giá trị của $|\log(LR)|$), một hiệu ứng thường được mong muốn trong khoa học pháp y. Vì lý do thứ hai này, các biến thể này của PAV phổ biến trong các tình huống mà điểm số huấn luyện khan hiếm.

### Hiệu suất của các phương pháp LR

Để đánh giá chứng cứ, các giá trị LR phải được tính toán từ các điểm số trong nhận dạng người nói pháp y bằng cách sử dụng các hệ hệ thống tự động hoặc bán tự động. Tuy nhiên, điều này là chưa đủ. LR tính toán được phải *hoạt động tốt*. Trong khung quyết định Bayes, điều này có nghĩa là LR tính toán được phải dẫn đến các xác suất cho phép đưa ra các quyết định tốt.

Ở cấp độ nguồn, việc đo lường hiệu suất thường được thực hiện theo cách thực nghiệm. Để đo lường hiệu suất của một phương pháp LR, một tập hợp *kiểm thử* gồm các giá trị LR cần được tạo ra bằng cách so sánh các đoạn tiếng nói từ một cơ sở dữ liệu *kiểm thử* của các danh tính đã biết, sử dụng phương pháp LR có hiệu suất cần đo lường. Các phép so sánh này thực chất là các vụ án mô phỏng, và do đó các điều kiện của tiếng nói phải tương tự như các điều kiện của tiếng nói chứng cứ trong kịch bản vụ án thực tế. Điều này dẫn đến một tập hợp các giá trị LR kiểm thử cùng nguồn (same-source) và các giá trị LR kiểm thử khác nguồn (different-source).

Một giải pháp để đo lường hiệu suất của các giá trị tỷ số khả tích đã được đề xuất cho nhận dạng người nói pháp y, và được đặt tên là **chi phí tỷ số khả tích log (log-likelihood-ratio cost)**, viết tắt là $C_{llr}$. Giá trị này là một trường hợp đặc biệt của hàm entropy chéo đã được đề cập trước đó. Do đó, $C_{llr}$ cũng có thể được sử dụng để tối ưu hóa các thuật toán tính toán LR phân biệt (chẳng hạn như hồi quy logistic).

$C_{llr}$ là một giá trị vô hướng có thể được tính toán dễ dàng từ một cơ sở dữ liệu gồm các giá trị LR kiểm thử, được tính toán như đã đề cập ở trên. Giá trị này càng cao thì hiệu suất của hệ thống nhận dạng người nói pháp y càng kém. Hơn nữa, có một giá trị *trần* là $C_{llr} = 1$ cho hiệu suất, và nếu một hệ thống nhận dạng người nói pháp y cho thấy $C_{llr} > 1$ trên một tập hợp các giá trị LR kiểm thử xác định, thì hệ thống đó không đáng tin cậy trong các điều kiện của các LR kiểm thử đó và phải bị loại bỏ. Nói cách khác, việc sử dụng các LR đó để tính toán tỷ số hậu nghiệm sẽ có xu hướng dẫn đến các quyết định kém hơn so với việc chỉ sử dụng duy nhất tỷ số tiên nghiệm.

Với các chỉ số hiệu suất thích hợp, có thể thực hiện quá trình xác thực (validation) phương pháp LR pháp y, được hiểu là quyết định về việc liệu một phương pháp nhận dạng người nói pháp y đã sẵn sàng để áp dụng vào thực tế hay chưa. Tuy nhiên, đây là một chủ đề rộng lớn nằm ngoài phạm vi của chương này.
